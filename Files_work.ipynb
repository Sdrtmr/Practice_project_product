```python
import pandas as pd
import sqlite3
from sqlite3 import Error
import os

def create_connection(db_file):
    """–°–æ–∑–¥–∞–Ω–∏–µ —Å–æ–µ–¥–∏–Ω–µ–Ω–∏—è —Å –±–∞–∑–æ–π –¥–∞–Ω–Ω—ã—Ö SQLite"""
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(f"‚úÖ –ü–æ–¥–∫–ª—é—á–µ–Ω–∏–µ –∫ SQLite —É—Å–ø–µ—à–Ω–æ: {db_file}")
        return conn
    except Error as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è: {e}")
    return conn

def create_tables(conn):
    """–°–æ–∑–¥–∞–Ω–∏–µ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã —Ç–∞–±–ª–∏—Ü –≤ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö"""
    try:
        cursor = conn.cursor()
        
        # –¢–∞–±–ª–∏—Ü–∞: –¢–∏–ø—ã –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤
        cursor.execute('''
        CREATE TABLE IF NOT EXISTS material_types (
            material_type TEXT PRIMARY KEY,
            loss_percentage REAL
        )
        ''')
        
        # –¢–∞–±–ª–∏—Ü–∞: –¢–∏–ø—ã –ø—Ä–æ–¥—É–∫—Ü–∏–∏
        cursor.execute('''
        CREATE TABLE IF NOT EXISTS product_types (
            product_type TEXT PRIMARY KEY,
            coefficient REAL
        )
        ''')
        
        # –¢–∞–±–ª–∏—Ü–∞: –¶–µ—Ö–∞
        cursor.execute('''
        CREATE TABLE IF NOT EXISTS workshops (
            workshop_name TEXT PRIMARY KEY,
            workshop_type TEXT,
            workers_count INTEGER
        )
        ''')
        
        # –¢–∞–±–ª–∏—Ü–∞: –ü—Ä–æ–¥—É–∫—Ü–∏—è
        cursor.execute('''
        CREATE TABLE IF NOT EXISTS products (
            product_id INTEGER PRIMARY KEY AUTOINCREMENT,
            product_type TEXT,
            product_name TEXT,
            article TEXT UNIQUE,
            min_partner_price REAL,
            main_material TEXT,
            FOREIGN KEY (product_type) REFERENCES product_types(product_type),
            FOREIGN KEY (main_material) REFERENCES material_types(material_type)
        )
        ''')
        
        # –¢–∞–±–ª–∏—Ü–∞: –ü—Ä–æ–¥—É–∫—Ü–∏—è –∏ —Ü–µ—Ö–∞ (—Å–≤—è–∑—å –º–Ω–æ–≥–∏–µ-–∫–æ-–º–Ω–æ–≥–∏–º)
        cursor.execute('''
        CREATE TABLE IF NOT EXISTS product_workshops (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            product_name TEXT,
            workshop_name TEXT,
            production_time REAL,
            FOREIGN KEY (product_name) REFERENCES products(product_name),
            FOREIGN KEY (workshop_name) REFERENCES workshops(workshop_name)
        )
        ''')
        
        conn.commit()
        print("‚úÖ –°—Ç—Ä—É–∫—Ç—É—Ä–∞ —Ç–∞–±–ª–∏—Ü —Å–æ–∑–¥–∞–Ω–∞ —É—Å–ø–µ—à–Ω–æ")
        
    except Error as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ —Ç–∞–±–ª–∏—Ü: {e}")

def import_material_types(conn, excel_file):
    """–ò–º–ø–æ—Ä—Ç —Ç–∏–ø–æ–≤ –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤"""
    try:
        df = pd.read_excel(excel_file)
        
        # –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã –¥–ª—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
        df.columns = ['material_type', 'loss_percentage']
        
        # –í—Å—Ç–∞–≤–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ç–∞–±–ª–∏—Ü—É
        df.to_sql('material_types', conn, if_exists='replace', index=False)
        
        print(f"‚úÖ –ò–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–æ {len(df)} —Ç–∏–ø–æ–≤ –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤")
        return True
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∏–º–ø–æ—Ä—Ç–µ —Ç–∏–ø–æ–≤ –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤: {e}")
        return False

def import_product_types(conn, excel_file):
    """–ò–º–ø–æ—Ä—Ç —Ç–∏–ø–æ–≤ –ø—Ä–æ–¥—É–∫—Ü–∏–∏"""
    try:
        df = pd.read_excel(excel_file)
        
        # –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã
        df.columns = ['product_type', 'coefficient']
        
        # –í—Å—Ç–∞–≤–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ç–∞–±–ª–∏—Ü—É
        df.to_sql('product_types', conn, if_exists='replace', index=False)
        
        print(f"‚úÖ –ò–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–æ {len(df)} —Ç–∏–ø–æ–≤ –ø—Ä–æ–¥—É–∫—Ü–∏–∏")
        return True
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∏–º–ø–æ—Ä—Ç–µ —Ç–∏–ø–æ–≤ –ø—Ä–æ–¥—É–∫—Ü–∏–∏: {e}")
        return False

def import_workshops(conn, excel_file):
    """–ò–º–ø–æ—Ä—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ü–µ—Ö–∞—Ö"""
    try:
        df = pd.read_excel(excel_file)
        
        # –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã
        df.columns = ['workshop_name', 'workshop_type', 'workers_count']
        
        # –í—Å—Ç–∞–≤–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ç–∞–±–ª–∏—Ü—É
        df.to_sql('workshops', conn, if_exists='replace', index=False)
        
        print(f"‚úÖ –ò–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–æ {len(df)} —Ü–µ—Ö–æ–≤")
        return True
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∏–º–ø–æ—Ä—Ç–µ —Ü–µ—Ö–æ–≤: {e}")
        return False

def import_products(conn, excel_file):
    """–ò–º–ø–æ—Ä—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –ø—Ä–æ–¥—É–∫—Ü–∏–∏"""
    try:
        df = pd.read_excel(excel_file)
        
        # –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã –¥–ª—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
        column_mapping = {
            '–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏': 'product_type',
            '–ù–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –ø—Ä–æ–¥—É–∫—Ü–∏–∏': 'product_name', 
            '–ê—Ä—Ç–∏–∫—É–ª': 'article',
            '–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞': 'min_partner_price',
            '–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Ç–µ—Ä–∏–∞–ª': 'main_material'
        }
        
        df = df.rename(columns=column_mapping)
        
        # –í—Å—Ç–∞–≤–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ç–∞–±–ª–∏—Ü—É
        df.to_sql('products', conn, if_exists='replace', index=False)
        
        print(f"‚úÖ –ò–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–æ {len(df)} –ø—Ä–æ–¥—É–∫—Ç–æ–≤")
        return True
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∏–º–ø–æ—Ä—Ç–µ –ø—Ä–æ–¥—É–∫—Ç–æ–≤: {e}")
        return False

def import_product_workshops(conn, excel_file):
    """–ò–º–ø–æ—Ä—Ç —Å–≤—è–∑–∏ –ø—Ä–æ–¥—É–∫—Ü–∏–∏ –∏ —Ü–µ—Ö–æ–≤"""
    try:
        df = pd.read_excel(excel_file)
        
        # –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã
        df.columns = ['product_name', 'workshop_name', 'production_time']
        
        # –í—Å—Ç–∞–≤–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Ç–∞–±–ª–∏—Ü—É
        df.to_sql('product_workshops', conn, if_exists='replace', index=False)
        
        print(f"‚úÖ –ò–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–æ {len(df)} —Å–≤—è–∑–µ–π –ø—Ä–æ–¥—É–∫—Ü–∏–∏ –∏ —Ü–µ—Ö–æ–≤")
        return True
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∏–º–ø–æ—Ä—Ç–µ —Å–≤—è–∑–µ–π: {e}")
        return False

def create_combined_view(conn):
    """–°–æ–∑–¥–∞–Ω–∏–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è —Å –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏"""
    try:
        cursor = conn.cursor()
        
        # –£–¥–∞–ª—è–µ–º —Å—Ç–∞—Ä–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –µ—Å–ª–∏ –æ–Ω–æ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
        cursor.execute("DROP VIEW IF EXISTS combined_data")
        
        # –°–æ–∑–¥–∞–µ–º –Ω–æ–≤–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ
        create_view_sql = '''
        CREATE VIEW combined_data AS
        SELECT 
            p.article as "–ê—Ä—Ç–∏–∫—É–ª",
            p.product_name as "–ù–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –ø—Ä–æ–¥—É–∫—Ü–∏–∏",
            p.product_type as "–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏",
            pt.coefficient as "–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —Ç–∏–ø–∞ –ø—Ä–æ–¥—É–∫—Ü–∏–∏",
            p.main_material as "–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Ç–µ—Ä–∏–∞–ª",
            mt.loss_percentage as "–ü—Ä–æ—Ü–µ–Ω—Ç –ø–æ—Ç–µ—Ä—å —Å—ã—Ä—å—è",
            p.min_partner_price as "–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞",
            pw.workshop_name as "–ù–∞–∑–≤–∞–Ω–∏–µ —Ü–µ—Ö–∞",
            pw.production_time as "–í—Ä–µ–º—è –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—è, —á",
            w.workshop_type as "–¢–∏–ø —Ü–µ—Ö–∞",
            w.workers_count as "–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–µ–ª–æ–≤–µ–∫ –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞"
        FROM products p
        LEFT JOIN product_types pt ON p.product_type = pt.product_type
        LEFT JOIN material_types mt ON p.main_material = mt.material_type
        LEFT JOIN product_workshops pw ON p.product_name = pw.product_name
        LEFT JOIN workshops w ON pw.workshop_name = w.workshop_name
        '''
        
        cursor.execute(create_view_sql)
        conn.commit()
        print("‚úÖ –ü—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ 'combined_data' —Å–æ–∑–¥–∞–Ω–æ —É—Å–ø–µ—à–Ω–æ")
        
    except Error as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è: {e}")

def show_database_stats(conn):
    """–ü–æ–∫–∞–∑ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö"""
    try:
        cursor = conn.cursor()
        
        print("\n" + "="*60)
        print("üìä –°–¢–ê–¢–ò–°–¢–ò–ö–ê –ë–ê–ó–´ –î–ê–ù–ù–´–•")
        print("="*60)
        
        # –°–ø–∏—Å–æ–∫ –≤—Å–µ—Ö —Ç–∞–±–ª–∏—Ü
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%'")
        tables = cursor.fetchall()
        
        print(f"üìã –í—Å–µ–≥–æ —Ç–∞–±–ª–∏—Ü: {len(tables)}")
        
        for table in tables:
            table_name = table[0]
            
            # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –≤ —Ç–∞–±–ª–∏—Ü–µ
            cursor.execute(f"SELECT COUNT(*) FROM {table_name}")
            row_count = cursor.fetchone()[0]
            
            # –°—Ç–æ–ª–±—Ü—ã —Ç–∞–±–ª–∏—Ü—ã
            cursor.execute(f"PRAGMA table_info({table_name})")
            columns = cursor.fetchall()
            
            print(f"\nüìÅ –¢–∞–±–ª–∏—Ü–∞: {table_name}")
            print(f"   üìà –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π: {row_count}")
            print(f"   üóÇÔ∏è  –°—Ç–æ–ª–±—Ü—ã ({len(columns)}):")
            
            for col in columns:
                col_name = col[1]
                col_type = col[2]
                print(f"      ‚Ä¢ {col_name} ({col_type})")
        
        # –ü—Ä–∏–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö –∏–∑ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è
        print("\n" + "="*60)
        print("üëÅÔ∏è –ü–†–ò–ú–ï–† –î–ê–ù–ù–´–• –ò–ó –û–ë–™–ï–î–ò–ù–ï–ù–ù–û–ì–û –ü–†–ï–î–°–¢–ê–í–õ–ï–ù–ò–Ø")
        print("="*60)
        
        cursor.execute("SELECT * FROM combined_data LIMIT 5")
        rows = cursor.fetchall()
        
        # –ü–æ–ª—É—á–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤
        cursor.execute("PRAGMA table_info(combined_data)")
        columns_info = cursor.fetchall()
        column_names = [col[1] for col in columns_info]
        
        print(f"\n–°—Ç–æ–ª–±—Ü—ã: {', '.join(column_names[:5])}...")
        print("\n–ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫:")
        
        for i, row in enumerate(rows, 1):
            print(f"\n{i}. " + "-" * 40)
            for col_name, value in zip(column_names, row):
                print(f"   {col_name}: {value}")
        
    except Error as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –ø–æ–ª—É—á–µ–Ω–∏–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏: {e}")

def run_sql_query(conn, query):
    """–í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ SQL-–∑–∞–ø—Ä–æ—Å–∞"""
    try:
        cursor = conn.cursor()
        cursor.execute(query)
        
        # –ü–æ–ª—É—á–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
        results = cursor.fetchall()
        
        # –ü–æ–ª—É—á–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤
        column_names = [description[0] for description in cursor.description]
        
        print(f"\nüìã –†–µ–∑—É–ª—å—Ç–∞—Ç –∑–∞–ø—Ä–æ—Å–∞ ({len(results)} —Å—Ç—Ä–æ–∫):")
        print("-" * 60)
        
        # –í—ã–≤–æ–¥–∏–º –∑–∞–≥–æ–ª–æ–≤–∫–∏ —Å—Ç–æ–ª–±—Ü–æ–≤
        print(" | ".join(column_names))
        print("-" * 60)
        
        # –í—ã–≤–æ–¥–∏–º –¥–∞–Ω–Ω—ã–µ
        for row in results[:10]:  # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º —Ç–æ–ª—å–∫–æ –ø–µ—Ä–≤—ã–µ 10 —Å—Ç—Ä–æ–∫
            print(" | ".join(str(value) for value in row))
            
        if len(results) > 10:
            print(f"... –∏ –µ—â–µ {len(results) - 10} —Å—Ç—Ä–æ–∫")
            
    except Error as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–ø—Ä–æ—Å–∞: {e}")

def main():
    """–û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è"""
    print("üöÄ –ó–ê–ü–£–°–ö –°–ö–†–ò–ü–¢–ê –ò–ú–ü–û–†–¢–ê –î–ê–ù–ù–´–• –ò–ó EXCEL –í SQLITE")
    print("="*60)
    
    # –ù–∞—Å—Ç—Ä–æ–π–∫–∏
    EXCEL_FILES = {
        'material_types': 'Material_type_import.xlsx',
        'product_types': 'Product_type_import.xlsx',
        'workshops': 'Workshops_import.xlsx',
        'products': 'Products_import.xlsx',
        'product_workshops': 'Product_workshops_import.xlsx'
    }
    
    DATABASE_FILE = "Big_base.db"
    
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤
    print("\nüîç –ü–†–û–í–ï–†–ö–ê –§–ê–ô–õ–û–í:")
    all_files_exist = True
    for file_type, file_path in EXCEL_FILES.items():
        if os.path.exists(file_path):
            print(f"   ‚úÖ {file_type}: {file_path}")
        else:
            print(f"   ‚ùå {file_type}: –§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω - {file_path}")
            all_files_exist = False
    
    if not all_files_exist:
        print("\n‚ö†Ô∏è –ù–µ–∫–æ—Ç–æ—Ä—ã–µ —Ñ–∞–π–ª—ã –Ω–µ –Ω–∞–π–¥–µ–Ω—ã! –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –ø—É—Ç–∏ –∫ —Ñ–∞–π–ª–∞–º.")
        return
    
    # –°–æ–∑–¥–∞–µ–º —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ —Å –±–∞–∑–æ–π –¥–∞–Ω–Ω—ã—Ö
    print(f"\nüîó –ü–û–î–ö–õ–Æ–ß–ï–ù–ò–ï –ö –ë–ê–ó–ï –î–ê–ù–ù–´–•:")
    conn = create_connection(DATABASE_FILE)
    if conn is None:
        return
    
    try:
        # –®–∞–≥ 1: –°–æ–∑–¥–∞–Ω–∏–µ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã —Ç–∞–±–ª–∏—Ü
        print("\nüîß –°–û–ó–î–ê–ù–ò–ï –°–¢–†–£–ö–¢–£–†–´ –¢–ê–ë–õ–ò–¶:")
        create_tables(conn)
        
        # –®–∞–≥ 2: –ò–º–ø–æ—Ä—Ç –¥–∞–Ω–Ω—ã—Ö
        print("\nüì• –ò–ú–ü–û–†–¢ –î–ê–ù–ù–´–• –ò–ó EXCEL:")
        
        # –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º —Ç–∏–ø—ã –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤
        print("\n1. –¢–∏–ø—ã –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤:")
        import_material_types(conn, EXCEL_FILES['material_types'])
        
        # –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º —Ç–∏–ø—ã –ø—Ä–æ–¥—É–∫—Ü–∏–∏
        print("\n2. –¢–∏–ø—ã –ø—Ä–æ–¥—É–∫—Ü–∏–∏:")
        import_product_types(conn, EXCEL_FILES['product_types'])
        
        # –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º —Ü–µ—Ö–∞
        print("\n3. –¶–µ—Ö–∞:")
        import_workshops(conn, EXCEL_FILES['workshops'])
        
        # –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –ø—Ä–æ–¥—É–∫—Ü–∏—é
        print("\n4. –ü—Ä–æ–¥—É–∫—Ü–∏—è:")
        import_products(conn, EXCEL_FILES['products'])
        
        # –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º —Å–≤—è–∑—å –ø—Ä–æ–¥—É–∫—Ü–∏–∏ –∏ —Ü–µ—Ö–æ–≤
        print("\n5. –°–≤—è–∑—å –ø—Ä–æ–¥—É–∫—Ü–∏–∏ –∏ —Ü–µ—Ö–æ–≤:")
        import_product_workshops(conn, EXCEL_FILES['product_workshops'])
        
        # –®–∞–≥ 3: –°–æ–∑–¥–∞–Ω–∏–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è
        print("\nüîó –°–û–ó–î–ê–ù–ò–ï –û–ë–™–ï–î–ò–ù–ï–ù–ù–û–ì–û –ü–†–ï–î–°–¢–ê–í–õ–ï–ù–ò–Ø:")
        create_combined_view(conn)
        
        # –®–∞–≥ 4: –ü–æ–∫–∞–∑ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
        show_database_stats(conn)
        
        # –®–∞–≥ 5: –ü—Ä–∏–º–µ—Ä—ã SQL-–∑–∞–ø—Ä–æ—Å–æ–≤
        print("\n" + "="*60)
        print("üí° –ü–†–ò–ú–ï–†–´ SQL-–ó–ê–ü–†–û–°–û–í")
        print("="*60)
        
        # –ü—Ä–∏–º–µ—Ä 1: –í—Å–µ –ø—Ä–æ–¥—É–∫—Ç—ã —Å –∏—Ö —Ç–∏–ø–∞–º–∏
        print("\nüìã –ü—Ä–∏–º–µ—Ä 1: –í—Å–µ –ø—Ä–æ–¥—É–∫—Ç—ã —Å –∏—Ö —Ç–∏–ø–∞–º–∏ –∏ –º–∞—Ç–µ—Ä–∏–∞–ª–∞–º–∏")
        query1 = """
        SELECT 
            p.article as "–ê—Ä—Ç–∏–∫—É–ª",
            p.product_name as "–ù–∞–∑–≤–∞–Ω–∏–µ",
            p.product_type as "–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏",
            pt.coefficient as "–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç",
            p.main_material as "–ú–∞—Ç–µ—Ä–∏–∞–ª",
            mt.loss_percentage as "–ü–æ—Ç–µ—Ä–∏ %"
        FROM products p
        JOIN product_types pt ON p.product_type = pt.product_type
        JOIN material_types mt ON p.main_material = mt.material_type
        LIMIT 10
        """
        run_sql_query(conn, query1)
        
        # –ü—Ä–∏–º–µ—Ä 2: –ü—Ä–æ–¥—É–∫—Ü–∏—è –ø–æ —Ü–µ—Ö–∞–º
        print("\nüìã –ü—Ä–∏–º–µ—Ä 2: –ü—Ä–æ–¥—É–∫—Ü–∏—è –ø–æ —Ü–µ—Ö–∞–º –∏ –≤—Ä–µ–º–µ–Ω–∏ –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—è")
        query2 = """
        SELECT 
            pw.workshop_name as "–¶–µ—Ö",
            w.workshop_type as "–¢–∏–ø —Ü–µ—Ö–∞",
            COUNT(pw.product_name) as "–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤",
            AVG(pw.production_time) as "–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è (—á)"
        FROM product_workshops pw
        JOIN workshops w ON pw.workshop_name = w.workshop_name
        GROUP BY pw.workshop_name
        ORDER BY "–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤" DESC
        """
        run_sql_query(conn, query2)
        
        # –ü—Ä–∏–º–µ—Ä 3: –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ü–∏–∏
        print("\nüìã –ü—Ä–∏–º–µ—Ä 3: –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ü–∏–∏")
        query3 = """
        SELECT 
            p.product_type as "–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏",
            COUNT(*) as "–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤",
            AVG(p.min_partner_price) as "–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞",
            MIN(p.min_partner_price) as "–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Ü–µ–Ω–∞",
            MAX(p.min_partner_price) as "–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Ü–µ–Ω–∞"
        FROM products p
        GROUP BY p.product_type
        ORDER BY "–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤" DESC
        """
        run_sql_query(conn, query3)
        
        # –ü—Ä–∏–º–µ—Ä 4: –ü–æ–ª–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–º –ø—Ä–æ–¥—É–∫—Ç–µ
        print("\nüìã –ü—Ä–∏–º–µ—Ä 4: –ü–æ–ª–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –ø—Ä–æ–¥—É–∫—Ç–µ (–∫—Ä–æ–≤–∞—Ç—å)")
        query4 = """
        SELECT * FROM combined_data 
        WHERE "–ù–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –ø—Ä–æ–¥—É–∫—Ü–∏–∏" LIKE '%–∫—Ä–æ–≤–∞—Ç—å%'
        LIMIT 5
        """
        run_sql_query(conn, query4)
        
        print("\n" + "="*60)
        print("üéâ –ò–ú–ü–û–†–¢ –î–ê–ù–ù–´–• –ó–ê–í–ï–†–®–ï–ù –£–°–ü–ï–®–ù–û!")
        print("="*60)
        print(f"\nüìÅ –ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö: {os.path.abspath(DATABASE_FILE)}")
        print("üìä –°–æ–∑–¥–∞–Ω–æ 5 –æ—Å–Ω–æ–≤–Ω—ã—Ö —Ç–∞–±–ª–∏—Ü –∏ 1 –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ")
        print("\nüí° –î–ª—è —Ä–∞–±–æ—Ç—ã —Å –¥–∞–Ω–Ω—ã–º–∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ:")
        print("   1. –¢–∞–±–ª–∏—Ü—ã: material_types, product_types, workshops, products, product_workshops")
        print("   2. –ü—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ: combined_data (–≤—Å–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ)")
        print("\nüîß –ü—Ä–∏–º–µ—Ä –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è –∫ –±–∞–∑–µ –≤ –¥—Ä—É–≥–æ–º —Å–∫—Ä–∏–ø—Ç–µ:")
        print(f"   conn = sqlite3.connect('{DATABASE_FILE}')")
        print("   cursor = conn.cursor()")
        print("   cursor.execute('SELECT * FROM combined_data LIMIT 5')")
        
    except Error as e:
        print(f"‚ùå –ö—Ä–∏—Ç–∏—á–µ—Å–∫–∞—è –æ—à–∏–±–∫–∞: {e}")
    finally:
        if conn:
            conn.close()
            print(f"\nüîí –°–æ–µ–¥–∏–Ω–µ–Ω–∏–µ —Å –±–∞–∑–æ–π –¥–∞–Ω–Ω—ã—Ö –∑–∞–∫—Ä—ã—Ç–æ")

if __name__ == "__main__":
    # –ó–∞–ø—É—Å–∫–∞–µ–º –æ—Å–Ω–æ–≤–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é
    main()
    
    # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è
    print("\n" + "="*60)
    print("üìö –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–ê–Ø –ò–ù–§–û–†–ú–ê–¶–ò–Ø")
    print("="*60)
    print("\n–î–ª—è –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç—ã:")
    print("1. DB Browser for SQLite (–±–µ—Å–ø–ª–∞—Ç–Ω–∞—è –ø—Ä–æ–≥—Ä–∞–º–º–∞ –¥–ª—è –ø—Ä–æ—Å–º–æ—Ç—Ä–∞ –ë–î)")
    print("2. Jupyter Notebook –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö —Å pandas")
    print("3. –õ—é–±–æ–π SQL-–∫–ª–∏–µ–Ω—Ç –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Å–ª–æ–∂–Ω—ã—Ö –∑–∞–ø—Ä–æ—Å–æ–≤")
    print("\n–ü—Ä–∏–º–µ—Ä –∞–Ω–∞–ª–∏–∑–∞ –≤ Python:")
    print('''
import sqlite3
import pandas as pd

# –ü–æ–¥–∫–ª—é—á–µ–Ω–∏–µ –∫ –±–∞–∑–µ
conn = sqlite3.connect('Big_base.db')

# –ß—Ç–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –≤ pandas DataFrame
df = pd.read_sql_query("SELECT * FROM combined_data", conn)

# –ê–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö
print(f"–í—Å–µ–≥–æ –∑–∞–ø–∏—Å–µ–π: {len(df)}")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤: {df['–ê—Ä—Ç–∏–∫—É–ª'].nunique()}")
print(f"–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞: {df['–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞'].mean():.2f}")

conn.close()
''')
```

```python
import sqlite3
import pandas as pd
from tabulate import tabulate
import textwrap

def display_full_table(database_path='Big_base.db', table_name='combined_data', 
                       max_rows=None, max_width=100, show_all_columns=True, 
                       columns_to_show=None, format_type='pretty'):
    """
    –í—ã–≤–æ–¥–∏—Ç –ø–æ–ª–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –∏–∑ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö SQLite –≤ —á–∏—Ç–∞–µ–º–æ–º —Ñ–æ—Ä–º–∞—Ç–µ
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    ----------
    database_path : str
        –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö SQLite
    table_name : str
        –ò–º—è —Ç–∞–±–ª–∏—Ü—ã –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 'combined_data')
    max_rows : int –∏–ª–∏ None
        –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è (None - –≤—Å–µ —Å—Ç—Ä–æ–∫–∏)
    max_width : int
        –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —à–∏—Ä–∏–Ω–∞ –≤—ã–≤–æ–¥–∞ –≤ —Å–∏–º–≤–æ–ª–∞—Ö
    show_all_columns : bool
        –ï—Å–ª–∏ True, –ø–æ–∫–∞–∑—ã–≤–∞—Ç—å –≤—Å–µ —Å—Ç–æ–ª–±—Ü—ã
    columns_to_show : list –∏–ª–∏ None
        –°–ø–∏—Å–æ–∫ —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è (–µ—Å–ª–∏ None –∏ show_all_columns=False, –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç—Å—è –∫–ª—é—á–µ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã)
    format_type : str
        –§–æ—Ä–º–∞—Ç –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è: 'pretty', 'grid', 'simple', 'github', 'html'
    """
    
    try:
        # –ü–æ–¥–∫–ª—é—á–µ–Ω–∏–µ –∫ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
        conn = sqlite3.connect(database_path)
        
        # –ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ —Ç–∞–±–ª–∏—Ü –≤ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
        cursor = conn.cursor()
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%'")
        tables = [table[0] for table in cursor.fetchall()]
        
        print("=" * 80)
        print(f"üìä –ë–ê–ó–ê –î–ê–ù–ù–´–•: {database_path}")
        print("=" * 80)
        print(f"–î–æ—Å—Ç—É–ø–Ω—ã–µ —Ç–∞–±–ª–∏—Ü—ã: {', '.join(tables)}")
        
        # –ï—Å–ª–∏ —Ç–∞–±–ª–∏—Ü–∞ –Ω–µ —É–∫–∞–∑–∞–Ω–∞, –ø–æ–∫–∞–∑—ã–≤–∞–µ–º —Å–ø–∏—Å–æ–∫ —Ç–∞–±–ª–∏—Ü
        if table_name is None:
            print("\nüìã –ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –¢–ê–ë–õ–ò–¶–ê–•:")
            for table in tables:
                cursor.execute(f"SELECT COUNT(*) FROM {table}")
                row_count = cursor.fetchone()[0]
                cursor.execute(f"PRAGMA table_info({table})")
                columns = cursor.fetchall()
                print(f"\n{table}:")
                print(f"  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫: {row_count}")
                print(f"  –°—Ç–æ–ª–±—Ü—ã ({len(columns)}): {', '.join([col[1] for col in columns])}")
            conn.close()
            return
        
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã
        if table_name not in tables:
            print(f"\n‚ùå –û–®–ò–ë–ö–ê: –¢–∞–±–ª–∏—Ü–∞ '{table_name}' –Ω–µ –Ω–∞–π–¥–µ–Ω–∞ –≤ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö!")
            print(f"–î–æ—Å—Ç—É–ø–Ω—ã–µ —Ç–∞–±–ª–∏—Ü—ã: {', '.join(tables)}")
            conn.close()
            return
        
        # –ü–æ–ª—É—á–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ç–∞–±–ª–∏—Ü–µ
        cursor.execute(f"PRAGMA table_info({table_name})")
        columns_info = cursor.fetchall()
        column_names = [col[1] for col in columns_info]
        
        # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –∫–∞–∫–∏–µ —Å—Ç–æ–ª–±—Ü—ã –ø–æ–∫–∞–∑—ã–≤–∞—Ç—å
        if not show_all_columns and columns_to_show is None:
            # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º —Ç–æ–ª—å–∫–æ –∫–ª—é—á–µ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã –¥–ª—è combined_data
            if table_name == 'combined_data':
                columns_to_show = ['–ê—Ä—Ç–∏–∫—É–ª', '–ù–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –ø—Ä–æ–¥—É–∫—Ü–∏–∏', '–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏', 
                                   '–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞', '–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Ç–µ—Ä–∏–∞–ª']
            else:
                columns_to_show = column_names[:min(5, len(column_names))]
        
        if columns_to_show is not None:
            # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ —É–∫–∞–∑–∞–Ω–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
            existing_columns = []
            for col in columns_to_show:
                if col in column_names:
                    existing_columns.append(col)
                else:
                    print(f"‚ö†Ô∏è –°—Ç–æ–ª–±–µ—Ü '{col}' –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ —Ç–∞–±–ª–∏—Ü–µ '{table_name}'")
            
            if not existing_columns:
                print(f"‚ùå –ù–µ—Ç —Å—É—â–µ—Å—Ç–≤—É—é—â–∏—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è!")
                conn.close()
                return
            
            column_names = existing_columns
        
        # –§–æ—Ä–º–∏—Ä—É–µ–º SQL –∑–∞–ø—Ä–æ—Å
        columns_str = ', '.join([f'"{col}"' for col in column_names])
        query = f"SELECT {columns_str} FROM {table_name}"
        
        # –î–æ–±–∞–≤–ª—è–µ–º –ª–∏–º–∏—Ç —Å—Ç—Ä–æ–∫ –µ—Å–ª–∏ —É–∫–∞–∑–∞–Ω
        if max_rows is not None:
            query += f" LIMIT {max_rows}"
        
        # –í—ã–ø–æ–ª–Ω—è–µ–º –∑–∞–ø—Ä–æ—Å
        df = pd.read_sql_query(query, conn)
        
        # –ó–∞–∫—Ä—ã–≤–∞–µ–º —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ
        conn.close()
        
        # –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
        total_rows = len(df)
        
        print(f"\nüìã –¢–ê–ë–õ–ò–¶–ê: {table_name}")
        print("-" * 80)
        print(f"–í—Å–µ–≥–æ —Å—Ç—Ä–æ–∫: {total_rows}")
        print(f"–û—Ç–æ–±—Ä–∞–∂–∞–µ–º—ã–µ —Å—Ç–æ–ª–±—Ü—ã ({len(column_names)}):")
        for i, col in enumerate(column_names, 1):
            print(f"  {i}. {col}")
        
        # –§–æ—Ä–º–∞—Ç–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∫—Ä–∞—Å–∏–≤–æ–≥–æ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
        if total_rows == 0:
            print("\n–¢–∞–±–ª–∏—Ü–∞ –ø—É—Å—Ç–∞!")
            return
        
        # –û–±—Ä–µ–∑–∞–µ–º –¥–ª–∏–Ω–Ω—ã–µ —Ç–µ–∫—Å—Ç–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
        df_display = df.copy()
        for col in df_display.columns:
            if df_display[col].dtype == 'object':  # –¢–µ–∫—Å—Ç–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã
                df_display[col] = df_display[col].apply(
                    lambda x: textwrap.shorten(str(x), width=30, placeholder='...') 
                    if isinstance(x, str) and len(str(x)) > 30 else x
                )
        
        # –í—ã–≤–æ–¥–∏–º —Ç–∞–±–ª–∏—Ü—É –≤ –≤—ã–±—Ä–∞–Ω–Ω–æ–º —Ñ–æ—Ä–º–∞—Ç–µ
        print("\n" + "=" * 80)
        print("üìä –î–ê–ù–ù–´–ï:")
        print("=" * 80)
        
        if format_type == 'html':
            # HTML —Ñ–æ—Ä–º–∞—Ç –¥–ª—è Jupyter Notebook
            display_html = df_display.to_html(max_cols=len(column_names), index=False)
            from IPython.display import HTML, display
            display(HTML(display_html))
        else:
            # –¢–µ–∫—Å—Ç–æ–≤—ã–µ —Ñ–æ—Ä–º–∞—Ç—ã
            tablefmt = format_type if format_type in ['pretty', 'grid', 'simple', 'github'] else 'grid'
            
            # –ò—Å–ø–æ–ª—å–∑—É–µ–º tabulate –¥–ª—è –∫—Ä–∞—Å–∏–≤–æ–≥–æ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
            table_str = tabulate(df_display, headers='keys', tablefmt=tablefmt, showindex=False)
            
            # –û–±—Ä–µ–∑–∞–µ–º —Å—Ç—Ä–æ–∫–∏ –µ—Å–ª–∏ –æ–Ω–∏ —Å–ª–∏—à–∫–æ–º –¥–ª–∏–Ω–Ω—ã–µ
            lines = table_str.split('\n')
            for line in lines[:100]:  # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –º–∞–∫—Å–∏–º—É–º 100 —Å—Ç—Ä–æ–∫ —Ç–∞–±–ª–∏—Ü—ã
                if len(line) > max_width:
                    print(line[:max_width] + '...')
                else:
                    print(line)
            
            if len(lines) > 100:
                print(f"\n... –∏ –µ—â–µ {len(lines) - 100} —Å—Ç—Ä–æ–∫")
        
        # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
        print("\n" + "=" * 80)
        print("üìà –°–¢–ê–¢–ò–°–¢–ò–ö–ê –ü–û –°–¢–û–õ–ë–¶–ê–ú:")
        print("=" * 80)
        
        for col in column_names:
            print(f"\nüìå {col}:")
            non_null = df[col].notna().sum()
            unique_count = df[col].nunique()
            print(f"   –ù–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {non_null}/{total_rows} ({non_null/total_rows*100:.1f}%)")
            print(f"   –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {unique_count}")
            
            # –î–ª—è —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –ø–æ–∫–∞–∑—ã–≤–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
            if pd.api.types.is_numeric_dtype(df[col]):
                print(f"   –ú–∏–Ω–∏–º—É–º: {df[col].min():.2f}")
                print(f"   –ú–∞–∫—Å–∏–º—É–º: {df[col].max():.2f}")
                print(f"   –°—Ä–µ–¥–Ω–µ–µ: {df[col].mean():.2f}")
                print(f"   –ú–µ–¥–∏–∞–Ω–∞: {df[col].median():.2f}")
            
            # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º —Ç–æ–ø-5 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
            if unique_count > 1:
                print(f"   –¢–æ–ø-5 —á–∞—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:")
                value_counts = df[col].value_counts().head(5)
                for value, count in value_counts.items():
                    percentage = count / total_rows * 100
                    print(f"     - {str(value)[:40]}: {count} ({percentage:.1f}%)")
        
        # –ü—Ä–∏–º–µ—Ä—ã SQL –∑–∞–ø—Ä–æ—Å–æ–≤ –¥–ª—è —ç—Ç–æ–π —Ç–∞–±–ª–∏—Ü—ã
        print("\n" + "=" * 80)
        print("üí° –ü–†–ò–ú–ï–†–´ SQL-–ó–ê–ü–†–û–°–û–í:")
        print("=" * 80)
        
        if table_name == 'combined_data':
            examples = [
                "1. –í—Å–µ –ø—Ä–æ–¥—É–∫—Ç—ã –¥–æ—Ä–æ–∂–µ 50000 —Ä—É–±–ª–µ–π:",
                f'   SELECT * FROM combined_data WHERE "–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞" > 50000',
                
                "\n2. –ü—Ä–æ–¥—É–∫—Ç—ã –∏–∑ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞:",
                f'   SELECT * FROM combined_data WHERE "–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Ç–µ—Ä–∏–∞–ª" LIKE "%–¥–µ—Ä–µ–≤–∞%"',
                
                "\n3. –ì—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –ø–æ —Ç–∏–ø—É –ø—Ä–æ–¥—É–∫—Ü–∏–∏:",
                f'   SELECT "–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏", COUNT(*) as –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ, AVG("–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞") as —Å—Ä–µ–¥–Ω—è—è_—Ü–µ–Ω–∞ FROM combined_data GROUP BY "–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏"',
                
                "\n4. –ü—Ä–æ–¥—É–∫—Ç—ã —Å –≤—Ä–µ–º–µ–Ω–µ–º –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—è –±–æ–ª—å—à–µ 3 —á–∞—Å–æ–≤:",
                f'   SELECT * FROM combined_data WHERE "–í—Ä–µ–º—è –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—è, —á" > 3',
                
                "\n5. –°–∞–º—ã–µ –¥–æ—Ä–æ–≥–∏–µ –ø—Ä–æ–¥—É–∫—Ç—ã (—Ç–æ–ø-10):",
                f'   SELECT * FROM combined_data ORDER BY "–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞" DESC LIMIT 10'
            ]
        else:
            examples = [
                "1. –í—Å–µ –∑–∞–ø–∏—Å–∏ –∏–∑ —Ç–∞–±–ª–∏—Ü—ã:",
                f'   SELECT * FROM {table_name}',
                
                "\n2. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π:",
                f'   SELECT COUNT(*) FROM {table_name}',
                
                "\n3. –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–µ—Ä–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞:",
                f'   SELECT DISTINCT "{column_names[0]}" FROM {table_name}',
            ]
        
        for example in examples:
            print(example)
        
        print("\n" + "=" * 80)
        print("‚úÖ –í–´–í–û–î –ó–ê–í–ï–†–®–ï–ù")
        print("=" * 80)
        
    except sqlite3.Error as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ SQLite: {e}")
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞: {e}")

# –§—É–Ω–∫—Ü–∏–∏-–ø–æ–º–æ—â–Ω–∏–∫–∏ –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
def show_combined_data(limit=20, show_all=False):
    """–ë—ã—Å—Ç—Ä—ã–π –≤—ã–≤–æ–¥ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö"""
    columns = None if show_all else ['–ê—Ä—Ç–∏–∫—É–ª', '–ù–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –ø—Ä–æ–¥—É–∫—Ü–∏–∏', '–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏', 
                                     '–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞', '–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Ç–µ—Ä–∏–∞–ª']
    display_full_table(table_name='combined_data', max_rows=limit, 
                       show_all_columns=show_all, columns_to_show=columns)

def show_table_info(database_path='Big_base.db'):
    """–ü–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –≤—Å–µ—Ö —Ç–∞–±–ª–∏—Ü–∞—Ö –≤ –±–∞–∑–µ"""
    display_full_table(database_path=database_path, table_name=None)

def export_to_excel(database_path='Big_base.db', table_name='combined_data', 
                   output_file='output.xlsx'):
    """–≠–∫—Å–ø–æ—Ä—Ç —Ç–∞–±–ª–∏—Ü—ã –≤ Excel"""
    try:
        conn = sqlite3.connect(database_path)
        query = f"SELECT * FROM {table_name}"
        df = pd.read_sql_query(query, conn)
        conn.close()
        
        df.to_excel(output_file, index=False)
        print(f"‚úÖ –î–∞–Ω–Ω—ã–µ —ç–∫—Å–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã –≤ {output_file}")
        print(f"   –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫: {len(df)}")
        print(f"   –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–æ–ª–±—Ü–æ–≤: {len(df.columns)}")
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ —ç–∫—Å–ø–æ—Ä—Ç–∞: {e}")

def run_custom_query(database_path='Big_base.db', query=None):
    """–í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ SQL-–∑–∞–ø—Ä–æ—Å–∞"""
    if query is None:
        print("–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, —É–∫–∞–∂–∏—Ç–µ SQL-–∑–∞–ø—Ä–æ—Å")
        return
    
    try:
        conn = sqlite3.connect(database_path)
        df = pd.read_sql_query(query, conn)
        conn.close()
        
        print(f"‚úÖ –ó–∞–ø—Ä–æ—Å –≤—ã–ø–æ–ª–Ω–µ–Ω —É—Å–ø–µ—à–Ω–æ")
        print(f"üìä –†–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤: {len(df)} —Å—Ç—Ä–æ–∫, {len(df.columns)} —Å—Ç–æ–ª–±—Ü–æ–≤")
        
        # –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü—ã
        print(tabulate(df.head(20), headers='keys', tablefmt='grid', showindex=False))
        
        if len(df) > 20:
            print(f"\n... –∏ –µ—â–µ {len(df) - 20} —Å—Ç—Ä–æ–∫")
        
        return df
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–ø—Ä–æ—Å–∞: {e}")

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
if __name__ == "__main__":
    print("üéØ –ò–ù–¢–ï–†–§–ï–ô–° –î–õ–Ø –†–ê–ë–û–¢–´ –° –ë–ê–ó–û–ô –î–ê–ù–ù–´–• Big_base.db")
    print("=" * 80)
    
    # –ü—Ä–∏–º–µ—Ä 1: –ü–æ–∫–∞–∑–∞—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –≤—Å–µ—Ö —Ç–∞–±–ª–∏—Ü–∞—Ö
    print("\n1. üìã –ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –í–°–ï–• –¢–ê–ë–õ–ò–¶–ê–•:")
    show_table_info()
    
    # –ü—Ä–∏–º–µ—Ä 2: –ü–æ–∫–∞–∑–∞—Ç—å –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ (–ø–µ—Ä–≤—ã–µ 10 —Å—Ç—Ä–æ–∫, –∫–ª—é—á–µ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã)
    print("\n\n2. üìä –û–ë–™–ï–î–ò–ù–ï–ù–ù–´–ï –î–ê–ù–ù–´–° (–ø–µ—Ä–≤—ã–µ 10 —Å—Ç—Ä–æ–∫):")
    show_combined_data(limit=10, show_all=False)
    
    # –ü—Ä–∏–º–µ—Ä 3: –ü–æ–∫–∞–∑–∞—Ç—å –ø–æ–ª–Ω—É—é —Ç–∞–±–ª–∏—Ü—É products
    print("\n\n3. üì¶ –¢–ê–ë–õ–ò–¶–ê PRODUCTS (–ø–µ—Ä–≤—ã–µ 15 —Å—Ç—Ä–æ–∫):")
    display_full_table(table_name='products', max_rows=15, format_type='grid')
    
    # –ü—Ä–∏–º–µ—Ä 4: –í—ã–ø–æ–ª–Ω–∏—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∑–∞–ø—Ä–æ—Å
    print("\n\n4. üîß –í–´–ü–û–õ–ù–ï–ù–ò–ï –ü–û–õ–¨–ó–û–í–ê–¢–ï–õ–¨–°–ö–û–ì–û –ó–ê–ü–†–û–°–ê:")
    custom_query = '''
    SELECT 
        "–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏",
        COUNT(*) as "–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤",
        AVG("–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞") as "–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞",
        MIN("–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞") as "–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Ü–µ–Ω–∞",
        MAX("–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è –ø–∞—Ä—Ç–Ω–µ—Ä–∞") as "–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Ü–µ–Ω–∞"
    FROM combined_data 
    GROUP BY "–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ü–∏–∏"
    ORDER BY "–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞" DESC
    '''
    
    result_df = run_custom_query(query=custom_query)
    
    # –ü—Ä–∏–º–µ—Ä 5: –≠–∫—Å–ø–æ—Ä—Ç –≤ Excel
    print("\n\n5. üíæ –≠–ö–°–ü–û–†–¢ –î–ê–ù–ù–´–• –í EXCEL:")
    export_to_excel(output_file='combined_data_export.xlsx')
    
    print("\n" + "=" * 80)
    print("üéâ –í–°–ï –û–ü–ï–†–ê–¶–ò–ò –í–´–ü–û–õ–ù–ï–ù–´ –£–°–ü–ï–®–ù–û!")
    print("=" * 80)
    
    print("\nüìö –ö–†–ê–¢–ö–ò–ô –°–ü–†–ê–í–û–ß–ù–ò–ö –ü–û –§–£–ù–ö–¶–ò–Ø–ú:")
    print("-" * 80)
    print("1. display_full_table() - –ü–æ–ª–Ω—ã–π –≤—ã–≤–æ–¥ –ª—é–±–æ–π —Ç–∞–±–ª–∏—Ü—ã —Å –Ω–∞—Å—Ç—Ä–æ–π–∫–∞–º–∏")
    print("2. show_combined_data() - –ë—ã—Å—Ç—Ä—ã–π –≤—ã–≤–æ–¥ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö")
    print("3. show_table_info() - –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –≤—Å–µ—Ö —Ç–∞–±–ª–∏—Ü–∞—Ö –≤ –±–∞–∑–µ")
    print("4. export_to_excel() - –≠–∫—Å–ø–æ—Ä—Ç —Ç–∞–±–ª–∏—Ü—ã –≤ Excel —Ñ–∞–π–ª")
    print("5. run_custom_query() - –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–≥–æ SQL-–∑–∞–ø—Ä–æ—Å–∞")
```

```python
import pandas as pd
import sqlite3
import json

def create_combined_table():
    """–°–æ–∑–¥–∞–µ—Ç –µ–¥–∏–Ω—É—é —Ç–∞–±–ª–∏—Ü—É, –æ–±—ä–µ–¥–∏–Ω—è—é—â—É—é –≤—Å–µ –¥–∞–Ω–Ω—ã–µ –∏–∑ –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤"""
    
    # 1. –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤
    df_workshops_time = pd.read_excel('Product_workshops_import_renamed.xlsx')
    df_material_types = pd.read_excel('Material_type_import_renamed.xlsx')
    df_product_types = pd.read_excel('Product_type_import_renamed.xlsx')
    df_products = pd.read_excel('Products_import_renamed.xlsx')
    df_workshops_info = pd.read_excel('Workshops_import_renamed.xlsx')
    
    # 2. –û–±—ä–µ–¥–∏–Ω—è–µ–º –¥–∞–Ω–Ω—ã–µ
    df_combined = df_workshops_time.copy()
    
    # –î–æ–±–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –ø—Ä–æ–¥—É–∫—Ç–∞—Ö
    product_info = df_products.set_index('Product Name')[['Product Type', 'Article', 'Minimum Partner Price', 'Main Material']]
    df_combined = df_combined.merge(product_info, left_on='Product Name', right_index=True, how='left')
    
    # –î–æ–±–∞–≤–ª—è–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Ç–∏–ø–æ–≤ –ø—Ä–æ–¥—É–∫—Ç–æ–≤
    product_type_coeff = df_product_types.set_index('Product Type')['Product Type Coefficient']
    df_combined['Product Type Coefficient'] = df_combined['Product Type'].map(product_type_coeff)
    
    # –î–æ–±–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –ø–æ—Ç–µ—Ä—è—Ö –º–∞—Ç–µ—Ä–∏–∞–ª–∞
    material_loss = df_material_types.set_index('Material Type')['Raw Material Loss Percentage']
    df_combined['Raw Material Loss Percentage'] = df_combined['Main Material'].map(material_loss)
    
    # –î–æ–±–∞–≤–ª—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ü–µ—Ö–∞—Ö
    workshops_info = df_workshops_info.set_index('Workshop Name')[['Workshop Type', 'Number of People for Production']]
    df_combined = df_combined.merge(workshops_info, left_on='Workshop Name', right_index=True, how='left')
    
    # 3. –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã
    column_mapping = {
        'Product Name': 'product_name',
        'Workshop Name': 'workshop_name',
        'Manufacturing Time (hours)': 'manufacturing_time_hours',
        'Product Type': 'product_type',
        'Article': 'article',
        'Minimum Partner Price': 'minimum_partner_price',
        'Main Material': 'main_material',
        'Product Type Coefficient': 'product_type_coefficient',
        'Raw Material Loss Percentage': 'raw_material_loss_percentage',
        'Workshop Type': 'workshop_type',
        'Number of People for Production': 'number_of_people_for_production'
    }
    df_combined.rename(columns=column_mapping, inplace=True)
    
    # 4. –î–æ–±–∞–≤–ª—è–µ–º –≤—ã—á–∏—Å–ª—è–µ–º—ã–µ —Å—Ç–æ–ª–±—Ü—ã
    df_combined['total_labor_hours'] = df_combined['manufacturing_time_hours'] * df_combined['number_of_people_for_production']
    
    # 5. –°–æ—Ä—Ç–∏—Ä—É–µ–º –∏ –¥–æ–±–∞–≤–ª—è–µ–º ID
    df_combined.sort_values(['product_name', 'workshop_name'], inplace=True)
    df_combined.reset_index(drop=True, inplace=True)
    df_combined['id'] = range(1, len(df_combined) + 1)
    
    # 6. –ü–µ—Ä–µ—É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã
    columns_order = [
        'id',
        'product_name',
        'article',
        'product_type',
        'product_type_coefficient',
        'minimum_partner_price',
        'main_material',
        'raw_material_loss_percentage',
        'workshop_name',
        'workshop_type',
        'number_of_people_for_production',
        'manufacturing_time_hours',
        'total_labor_hours'
    ]
    
    return df_combined[columns_order]

def save_all_formats(df_combined):
    """–°–æ—Ö—Ä–∞–Ω—è–µ—Ç –¥–∞–Ω–Ω—ã–µ –≤–æ –≤—Å–µ—Ö —Ñ–æ—Ä–º–∞—Ç–∞—Ö"""
    # Excel
    df_combined.to_excel('combined_data.xlsx', index=False)
    print("‚úì Excel —Ñ–∞–π–ª —Å–æ–∑–¥–∞–Ω: combined_data.xlsx")
    
    # SQLite
    conn = sqlite3.connect('furniture_production.db')
    df_combined.to_sql('production_data', conn, if_exists='replace', index=False)
    
    # –°–æ–∑–¥–∞–µ–º –∏–Ω–¥–µ–∫—Å—ã
    cursor = conn.cursor()
    cursor.execute('CREATE INDEX idx_product ON production_data(product_name)')
    cursor.execute('CREATE INDEX idx_workshop ON production_data(workshop_name)')
    conn.commit()
    conn.close()
    print("‚úì –ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö SQLite —Å–æ–∑–¥–∞–Ω–∞: furniture_production.db")
    
    # CSV
    df_combined.to_csv('combined_data.csv', index=False, encoding='utf-8-sig')
    print("‚úì CSV —Ñ–∞–π–ª —Å–æ–∑–¥–∞–Ω: combined_data.csv")
    
    # JSON
    data_dict = df_combined.to_dict(orient='records')
    with open('combined_data.json', 'w', encoding='utf-8') as f:
        json.dump(data_dict, f, ensure_ascii=False, indent=2)
    print("‚úì JSON —Ñ–∞–π–ª —Å–æ–∑–¥–∞–Ω: combined_data.json")
    
    # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –æ—Ç—á–µ—Ç—ã
    product_summary = df_combined.groupby(['product_name', 'product_type']).agg({
        'manufacturing_time_hours': 'sum',
        'minimum_partner_price': 'first',
        'article': 'first'
    }).reset_index()
    product_summary.to_excel('product_summary.xlsx', index=False)
    print("‚úì –°–≤–æ–¥–Ω—ã–π –æ—Ç—á–µ—Ç –ø–æ –ø—Ä–æ–¥—É–∫—Ç–∞–º: product_summary.xlsx")

def main():
    """–û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è"""
    print("–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –∏–∑ –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤...")
    print("-" * 50)
    
    try:
        # –°–æ–∑–¥–∞–µ–º –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—É—é —Ç–∞–±–ª–∏—Ü—É
        df_combined = create_combined_table()
        
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ç–∞–±–ª–∏—Ü–µ
        print(f"–°–æ–∑–¥–∞–Ω–∞ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ —Å {len(df_combined)} —Å—Ç—Ä–æ–∫–∞–º–∏")
        print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–æ–ª–±—Ü–æ–≤: {len(df_combined.columns)}")
        print("\n–°—Ç—Ä—É–∫—Ç—É—Ä–∞ —Ç–∞–±–ª–∏—Ü—ã:")
        for col in df_combined.columns:
            print(f"  - {col}")
        
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ —Å—Ç—Ä–æ–∫–∏
        print("\n–ü–µ—Ä–≤—ã–µ 3 —Å—Ç—Ä–æ–∫–∏ –¥–∞–Ω–Ω—ã—Ö:")
        print(df_combined.head(3).to_string(index=False))
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤–æ –≤—Å–µ—Ö —Ñ–æ—Ä–º–∞—Ç–∞—Ö
        print("\n" + "-" * 50)
        print("–°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö...")
        save_all_formats(df_combined)
        
        print("\n" + "-" * 50)
        print("–í–´–ü–û–õ–ù–ï–ù–û –£–°–ü–ï–®–ù–û!")
        print("\n–°–æ–∑–¥–∞–Ω—ã —Ñ–∞–π–ª—ã:")
        print("1. combined_data.xlsx - –û—Å–Ω–æ–≤–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –≤ Excel")
        print("2. furniture_production.db - –ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö SQLite")
        print("3. combined_data.csv - –î–∞–Ω–Ω—ã–µ –≤ CSV —Ñ–æ—Ä–º–∞—Ç–µ")
        print("4. combined_data.json - –î–∞–Ω–Ω—ã–µ –≤ JSON —Ñ–æ—Ä–º–∞—Ç–µ")
        print("5. product_summary.xlsx - –°–≤–æ–¥–Ω—ã–π –æ—Ç—á–µ—Ç –ø–æ –ø—Ä–æ–¥—É–∫—Ç–∞–º")
        
        return df_combined
        
    except FileNotFoundError as e:
        print(f"–û–®–ò–ë–ö–ê: –ù–µ –Ω–∞–π–¥–µ–Ω —Ñ–∞–π–ª {e}")
        print("\n–£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –≤—Å–µ —Ñ–∞–π–ª—ã –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ —Ç–æ–π –∂–µ –ø–∞–ø–∫–µ:")
        print("1. Product_workshops_import_renamed.xlsx")
        print("2. Material_type_import_renamed.xlsx")
        print("3. Product_type_import_renamed.xlsx")
        print("4. Products_import_renamed.xlsx")
        print("5. Workshops_import_renamed.xlsx")
        return None
    except Exception as e:
        print(f"–û–®–ò–ë–ö–ê: {e}")
        return None

if __name__ == "__main__":
    main()
```

```python
import sqlite3
import pandas as pd
from datetime import datetime
import os

def export_database_to_excel(db_path='furniture_production.db', output_path=None):
    """
    –≠–∫—Å–ø–æ—Ä—Ç–∏—Ä—É–µ—Ç –≤—Å–µ —Ç–∞–±–ª–∏—Ü—ã –∏–∑ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö SQLite –≤ Excel —Ñ–∞–π–ª —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –ª–∏—Å—Ç–∞–º–∏
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    ----------
    db_path : str
        –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö SQLite
    output_path : str, optional
        –ü—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è Excel —Ñ–∞–π–ª–∞. –ï—Å–ª–∏ –Ω–µ —É–∫–∞–∑–∞–Ω, –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç—Å—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏.
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    -----------
    str
        –ü—É—Ç—å –∫ —Å–æ–∑–¥–∞–Ω–Ω–æ–º—É Excel —Ñ–∞–π–ª—É
    """
    
    try:
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
        if not os.path.exists(db_path):
            raise FileNotFoundError(f"–ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö –Ω–µ –Ω–∞–π–¥–µ–Ω–∞: {db_path}")
        
        # –ü–æ–¥–∫–ª—é—á–∞–µ–º—Å—è –∫ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
        conn = sqlite3.connect(db_path)
        
        # –ü–æ–ª—É—á–∞–µ–º —Å–ø–∏—Å–æ–∫ –≤—Å–µ—Ö —Ç–∞–±–ª–∏—Ü –≤ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
        cursor = conn.cursor()
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
        tables = cursor.fetchall()
        
        if not tables:
            raise ValueError(f"–ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö {db_path} –Ω–µ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç–∞–±–ª–∏—Ü")
        
        table_names = [table[0] for table in tables]
        
        print(f"–ù–∞–π–¥–µ–Ω–æ —Ç–∞–±–ª–∏—Ü –≤ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö: {len(table_names)}")
        print(f"–¢–∞–±–ª–∏—Ü—ã: {', '.join(table_names)}")
        
        # –ï—Å–ª–∏ –ø—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –Ω–µ —É–∫–∞–∑–∞–Ω, –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏
        if output_path is None:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            output_path = f"database_export_{timestamp}.xlsx"
        
        # –°–æ–∑–¥–∞–µ–º Excel writer –¥–ª—è –∑–∞–ø–∏—Å–∏ –≤ –Ω–µ—Å–∫–æ–ª—å–∫–æ –ª–∏—Å—Ç–æ–≤
        with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
            for table_name in table_names:
                # –ß–∏—Ç–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ —Ç–∞–±–ª–∏—Ü—ã
                query = f"SELECT * FROM {table_name}"
                df = pd.read_sql_query(query, conn)
                
                # –ó–∞–ø–∏—Å—ã–≤–∞–µ–º –≤ Excel
                sheet_name = table_name[:31]  # Excel –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ—Ç –∏–º–µ–Ω–∞ –ª–∏—Å—Ç–æ–≤ 31 —Å–∏–º–≤–æ–ª–æ–º
                df.to_excel(writer, sheet_name=sheet_name, index=False)
                
                print(f"  –¢–∞–±–ª–∏—Ü–∞ '{table_name}' —ç–∫—Å–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–∞ ({len(df)} —Å—Ç—Ä–æ–∫)")
        
        # –ó–∞–∫—Ä—ã–≤–∞–µ–º —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ
        conn.close()
        
        print(f"\n‚úì –ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö —É—Å–ø–µ—à–Ω–æ —ç–∫—Å–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–∞ –≤ —Ñ–∞–π–ª: {output_path}")
        print(f"  –†–∞–∑–º–µ—Ä —Ñ–∞–π–ª–∞: {os.path.getsize(output_path) / 1024:.2f} KB")
        
        return output_path
    
    except sqlite3.Error as e:
        print(f"–û—à–∏–±–∫–∞ SQLite: {e}")
        raise
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ —ç–∫—Å–ø–æ—Ä—Ç–µ –¥–∞–Ω–Ω—ã—Ö: {e}")
        raise


def export_with_summary_reports(db_path='furniture_production.db', output_path=None):
    """
    –≠–∫—Å–ø–æ—Ä—Ç–∏—Ä—É–µ—Ç –±–∞–∑—É –¥–∞–Ω–Ω—ã—Ö –≤ Excel —Å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–º–∏ —Å–≤–æ–¥–Ω—ã–º–∏ –æ—Ç—á–µ—Ç–∞–º–∏
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    ----------
    db_path : str
        –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö SQLite
    output_path : str, optional
        –ü—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è Excel —Ñ–∞–π–ª–∞
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    -----------
    str
        –ü—É—Ç—å –∫ —Å–æ–∑–¥–∞–Ω–Ω–æ–º—É Excel —Ñ–∞–π–ª—É
    """
    
    try:
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
        if not os.path.exists(db_path):
            raise FileNotFoundError(f"–ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö –Ω–µ –Ω–∞–π–¥–µ–Ω–∞: {db_path}")
        
        # –ü–æ–¥–∫–ª—é—á–∞–µ–º—Å—è –∫ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
        conn = sqlite3.connect(db_path)
        
        # –ï—Å–ª–∏ –ø—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –Ω–µ —É–∫–∞–∑–∞–Ω, –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏
        if output_path is None:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            output_path = f"furniture_database_full_export_{timestamp}.xlsx"
        
        # –°–æ–∑–¥–∞–µ–º Excel writer
        with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
            print("–≠–∫—Å–ø–æ—Ä—Ç –¥–∞–Ω–Ω—ã—Ö –∏–∑ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö...")
            
            # 1. –≠–∫—Å–ø–æ—Ä—Ç–∏—Ä—É–µ–º –≤—Å–µ —Ç–∞–±–ª–∏—Ü—ã –∏–∑ –±–∞–∑—ã
            cursor = conn.cursor()
            cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
            tables = [table[0] for table in cursor.fetchall()]
            
            for table_name in tables:
                df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
                sheet_name = table_name[:31]
                df.to_excel(writer, sheet_name=sheet_name, index=False)
                print(f"  ‚úì –¢–∞–±–ª–∏—Ü–∞ '{table_name}' —ç–∫—Å–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–∞")
            
            # 2. –°–æ–∑–¥–∞–µ–º —Å–≤–æ–¥–Ω—ã–µ –æ—Ç—á–µ—Ç—ã (–µ—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ç–∞–±–ª–∏—Ü–∞ production_data)
            if 'production_data' in tables:
                print("\n–°–æ–∑–¥–∞–Ω–∏–µ —Å–≤–æ–¥–Ω—ã—Ö –æ—Ç—á–µ—Ç–æ–≤...")
                
                # –ó–∞–≥—Ä—É–∂–∞–µ–º –æ—Å–Ω–æ–≤–Ω—É—é —Ç–∞–±–ª–∏—Ü—É
                df_production = pd.read_sql_query("SELECT * FROM production_data", conn)
                
                # 2.1. –°–≤–æ–¥–∫–∞ –ø–æ –ø—Ä–æ–¥—É–∫—Ç–∞–º
                product_summary = df_production.groupby(['product_name', 'product_type']).agg({
                    'article': 'first',
                    'minimum_partner_price': 'first',
                    'main_material': 'first',
                    'manufacturing_time_hours': 'sum',
                    'total_labor_hours': 'sum'
                }).reset_index()
                
                product_summary = product_summary.sort_values('manufacturing_time_hours', ascending=False)
                product_summary.to_excel(writer, sheet_name='Product_Summary', index=False)
                print(f"  ‚úì –°–≤–æ–¥–∫–∞ –ø–æ –ø—Ä–æ–¥—É–∫—Ç–∞–º —Å–æ–∑–¥–∞–Ω–∞ ({len(product_summary)} –ø—Ä–æ–¥—É–∫—Ç–æ–≤)")
                
                # 2.2. –°–≤–æ–¥–∫–∞ –ø–æ —Ü–µ—Ö–∞–º
                workshop_summary = df_production.groupby(['workshop_name', 'workshop_type']).agg({
                    'number_of_people_for_production': 'first',
                    'manufacturing_time_hours': 'sum',
                    'total_labor_hours': 'sum'
                }).reset_index()
                
                workshop_summary = workshop_summary.sort_values('total_labor_hours', ascending=False)
                workshop_summary.to_excel(writer, sheet_name='Workshop_Summary', index=False)
                print(f"  ‚úì –°–≤–æ–¥–∫–∞ –ø–æ —Ü–µ—Ö–∞–º —Å–æ–∑–¥–∞–Ω–∞ ({len(workshop_summary)} —Ü–µ—Ö–æ–≤)")
                
                # 2.3. –ú–∞—Ç—Ä–∏—Ü–∞ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ (–ø—Ä–æ–¥—É–∫—Ç—ã √ó —Ü–µ—Ö–∞)
                time_matrix = pd.pivot_table(
                    df_production,
                    values='manufacturing_time_hours',
                    index='product_name',
                    columns='workshop_name',
                    aggfunc='sum',
                    fill_value=0
                )
                time_matrix.to_excel(writer, sheet_name='Time_Matrix')
                print(f"  ‚úì –ú–∞—Ç—Ä–∏—Ü–∞ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ —Å–æ–∑–¥–∞–Ω–∞")
                
                # 2.4. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ç–æ–≤
                product_type_stats = df_production.groupby('product_type').agg({
                    'product_name': pd.Series.nunique,
                    'manufacturing_time_hours': 'sum',
                    'total_labor_hours': 'sum',
                    'minimum_partner_price': 'mean'
                }).rename(columns={
                    'product_name': 'number_of_products',
                    'minimum_partner_price': 'avg_price'
                }).reset_index()
                
                product_type_stats = product_type_stats.sort_values('total_labor_hours', ascending=False)
                product_type_stats.to_excel(writer, sheet_name='ProductType_Stats', index=False)
                print(f"  ‚úì –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ç–æ–≤ —Å–æ–∑–¥–∞–Ω–∞")
                
                # 2.5. –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞–º
                material_stats = df_production.groupby('main_material').agg({
                    'product_name': pd.Series.nunique,
                    'manufacturing_time_hours': 'sum',
                    'raw_material_loss_percentage': 'first'
                }).rename(columns={
                    'product_name': 'number_of_products'
                }).reset_index()
                
                material_stats = material_stats.sort_values('number_of_products', ascending=False)
                material_stats.to_excel(writer, sheet_name='Material_Stats', index=False)
                print(f"  ‚úì –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞–º —Å–æ–∑–¥–∞–Ω–∞")
            
            # 3. –°–æ–∑–¥–∞–µ–º –ª–∏—Å—Ç —Å –º–µ—Ç–∞–¥–∞–Ω–Ω—ã–º–∏
            metadata = {
                '–≠–∫—Å–ø–æ—Ä—Ç —Å–æ–∑–¥–∞–Ω': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                '–ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö': db_path,
                '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–∞–±–ª–∏—Ü': len(tables),
                '–¢–∞–±–ª–∏—Ü—ã': ', '.join(tables),
                '–†–∞–∑–º–µ—Ä —Ñ–∞–π–ª–∞ –ë–î': f"{os.path.getsize(db_path) / 1024:.2f} KB"
            }
            
            df_metadata = pd.DataFrame(list(metadata.items()), columns=['–ü–∞—Ä–∞–º–µ—Ç—Ä', '–ó–Ω–∞—á–µ–Ω–∏–µ'])
            df_metadata.to_excel(writer, sheet_name='Metadata', index=False)
            print(f"  ‚úì –ú–µ—Ç–∞–¥–∞–Ω–Ω—ã–µ –¥–æ–±–∞–≤–ª–µ–Ω—ã")
        
        # –ó–∞–∫—Ä—ã–≤–∞–µ–º —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ
        conn.close()
        
        # –í—ã–≤–æ–¥–∏–º –∏—Ç–æ–≥–æ–≤—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
        file_size = os.path.getsize(output_path) / 1024 / 1024  # –≤ MB
        print(f"\n{'='*60}")
        print(f"‚úì –≠–ö–°–ü–û–†–¢ –£–°–ü–ï–®–ù–û –ó–ê–í–ï–†–®–ï–ù!")
        print(f"{'='*60}")
        print(f"–§–∞–π–ª: {output_path}")
        print(f"–†–∞–∑–º–µ—Ä: {file_size:.2f} MB")
        print(f"–î–∞—Ç–∞ —Å–æ–∑–¥–∞–Ω–∏—è: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        
        return output_path
    
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ —ç–∫—Å–ø–æ—Ä—Ç–µ —Å –æ—Ç—á–µ—Ç–∞–º–∏: {e}")
        if 'conn' in locals():
            conn.close()
        raise


def export_custom_query(db_path='furniture_production.db', query=None, output_path=None, sheet_name='Query_Result'):
    """
    –≠–∫—Å–ø–æ—Ä—Ç–∏—Ä—É–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç SQL –∑–∞–ø—Ä–æ—Å–∞ –≤ Excel —Ñ–∞–π–ª
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    ----------
    db_path : str
        –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö SQLite
    query : str
        SQL –∑–∞–ø—Ä–æ—Å –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è
    output_path : str, optional
        –ü—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è Excel —Ñ–∞–π–ª–∞
    sheet_name : str, optional
        –ò–º—è –ª–∏—Å—Ç–∞ –≤ Excel —Ñ–∞–π–ª–µ
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    -----------
    str
        –ü—É—Ç—å –∫ —Å–æ–∑–¥–∞–Ω–Ω–æ–º—É Excel —Ñ–∞–π–ª—É
    """
    
    try:
        if query is None:
            # –ü—Ä–∏–º–µ—Ä –∑–∞–ø—Ä–æ—Å–∞ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
            query = """
            SELECT 
                p.product_name,
                p.product_type,
                p.article,
                p.minimum_partner_price,
                p.main_material,
                SUM(pd.manufacturing_time_hours) as total_production_hours,
                SUM(pd.total_labor_hours) as total_labor_hours,
                COUNT(DISTINCT pd.workshop_name) as workshops_count
            FROM production_data pd
            LEFT JOIN (
                SELECT DISTINCT product_name, product_type, article, minimum_partner_price, main_material 
                FROM production_data
            ) p ON pd.product_name = p.product_name
            GROUP BY pd.product_name
            ORDER BY total_production_hours DESC
            """
        
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
        if not os.path.exists(db_path):
            raise FileNotFoundError(f"–ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö –Ω–µ –Ω–∞–π–¥–µ–Ω–∞: {db_path}")
        
        # –ü–æ–¥–∫–ª—é—á–∞–µ–º—Å—è –∫ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
        conn = sqlite3.connect(db_path)
        
        # –í—ã–ø–æ–ª–Ω—è–µ–º –∑–∞–ø—Ä–æ—Å
        df_result = pd.read_sql_query(query, conn)
        
        # –ó–∞–∫—Ä—ã–≤–∞–µ–º —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ
        conn.close()
        
        # –ï—Å–ª–∏ –ø—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –Ω–µ —É–∫–∞–∑–∞–Ω, –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏
        if output_path is None:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            output_path = f"query_result_{timestamp}.xlsx"
        
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ Excel
        df_result.to_excel(output_path, sheet_name=sheet_name[:31], index=False)
        
        print(f"‚úì –†–µ–∑—É–ª—å—Ç–∞—Ç –∑–∞–ø—Ä–æ—Å–∞ —ç–∫—Å–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω –≤ —Ñ–∞–π–ª: {output_path}")
        print(f"  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫: {len(df_result)}")
        print(f"  –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–æ–ª–±—Ü–æ–≤: {len(df_result.columns)}")
        
        return output_path
    
    except sqlite3.Error as e:
        print(f"–û—à–∏–±–∫–∞ SQLite –≤ –∑–∞–ø—Ä–æ—Å–µ: {e}")
        raise
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ –∑–∞–ø—Ä–æ—Å–∞: {e}")
        raise


def create_detailed_report(db_path='furniture_production.db'):
    """
    –°–æ–∑–¥–∞–µ—Ç –¥–µ—Ç–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –æ—Ç—á–µ—Ç —Å –∞–Ω–∞–ª–∏–∑–æ–º –¥–∞–Ω–Ω—ã—Ö
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    ----------
    db_path : str
        –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö SQLite
    """
    
    try:
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
        if not os.path.exists(db_path):
            raise FileNotFoundError(f"–ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö –Ω–µ –Ω–∞–π–¥–µ–Ω–∞: {db_path}")
        
        # –ü–æ–¥–∫–ª—é—á–∞–µ–º—Å—è –∫ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
        conn = sqlite3.connect(db_path)
        
        # –°–æ–∑–¥–∞–µ–º –∏–º—è —Ñ–∞–π–ª–∞ —Å timestamp
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        output_path = f"furniture_detailed_report_{timestamp}.xlsx"
        
        print(f"–°–æ–∑–¥–∞–Ω–∏–µ –¥–µ—Ç–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –æ—Ç—á–µ—Ç–∞...")
        print(f"–§–∞–π–ª –æ—Ç—á–µ—Ç–∞: {output_path}")
        
        # –°–æ–∑–¥–∞–µ–º Excel writer
        with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
            
            # –ó–∞–≥—Ä—É–∂–∞–µ–º –æ—Å–Ω–æ–≤–Ω—É—é —Ç–∞–±–ª–∏—Ü—É
            df = pd.read_sql_query("SELECT * FROM production_data", conn)
            
            # 1. –û–±—â–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –æ–±—â–µ–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏...")
            total_stats = pd.DataFrame({
                '–ú–µ—Ç—Ä–∏–∫–∞': [
                    '–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π',
                    '–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤',
                    '–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ü–µ—Ö–æ–≤',
                    '–û–±—â–µ–µ –≤—Ä–µ–º—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ (—á–∞—Å—ã)',
                    '–û–±—â–∏–µ —Ç—Ä—É–¥–æ–∑–∞—Ç—Ä–∞—Ç—ã (—á–µ–ª–æ–≤–µ–∫–æ-—á–∞—Å—ã)',
                    '–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –Ω–∞ –ø—Ä–æ–¥—É–∫—Ç (—á–∞—Å—ã)',
                    '–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –ø—Ä–æ–¥—É–∫—Ç–∞',
                    '–î–∞—Ç–∞ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –æ—Ç—á–µ—Ç–∞'
                ],
                '–ó–Ω–∞—á–µ–Ω–∏–µ': [
                    len(df),
                    df['product_name'].nunique(),
                    df['workshop_name'].nunique(),
                    df['manufacturing_time_hours'].sum(),
                    df['total_labor_hours'].sum(),
                    df.groupby('product_name')['manufacturing_time_hours'].sum().mean(),
                    df['minimum_partner_price'].mean(),
                    datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                ]
            })
            total_stats.to_excel(writer, sheet_name='–û–±—â–∞—è_—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞', index=False)
            
            # 2. –î–µ—Ç–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –ø—Ä–æ–¥—É–∫—Ç–æ–≤
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –¥–µ—Ç–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã –ø—Ä–æ–¥—É–∫—Ç–æ–≤...")
            product_details = df.groupby('product_name').agg({
                'article': 'first',
                'product_type': 'first',
                'product_type_coefficient': 'first',
                'minimum_partner_price': 'first',
                'main_material': 'first',
                'raw_material_loss_percentage': 'first',
                'manufacturing_time_hours': 'sum',
                'total_labor_hours': 'sum'
            }).reset_index()
            
            product_details = product_details.sort_values('manufacturing_time_hours', ascending=False)
            product_details.to_excel(writer, sheet_name='–ü—Ä–æ–¥—É–∫—Ç—ã_–¥–µ—Ç–∞–ª—å–Ω–æ', index=False)
            
            # 3. –ê–Ω–∞–ª–∏–∑ –ø–æ —Ü–µ—Ö–∞–º
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –∞–Ω–∞–ª–∏–∑–∞ –ø–æ —Ü–µ—Ö–∞–º...")
            workshop_analysis = df.groupby(['workshop_name', 'workshop_type']).agg({
                'number_of_people_for_production': 'first',
                'manufacturing_time_hours': ['sum', 'mean', 'count'],
                'total_labor_hours': 'sum'
            }).reset_index()
            
            workshop_analysis.columns = [
                'workshop_name', 'workshop_type', 'number_of_people',
                'total_time_sum', 'avg_time_per_record', 'record_count', 'total_labor_hours'
            ]
            
            workshop_analysis = workshop_analysis.sort_values('total_labor_hours', ascending=False)
            workshop_analysis.to_excel(writer, sheet_name='–ê–Ω–∞–ª–∏–∑_—Ü–µ—Ö–æ–≤', index=False)
            
            # 4. –¢–æ–ø-10 —Å–∞–º—ã—Ö —Ç—Ä—É–¥–æ–µ–º–∫–∏—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤
            print("  –°–æ–∑–¥–∞–Ω–∏–µ —Ç–æ–ø-10 —Å–∞–º—ã—Ö —Ç—Ä—É–¥–æ–µ–º–∫–∏—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤...")
            top_products = product_details.nlargest(10, 'total_labor_hours')[['product_name', 'product_type', 'total_labor_hours', 'manufacturing_time_hours']]
            top_products.to_excel(writer, sheet_name='–¢–æ–ø10_–ø—Ä–æ–¥—É–∫—Ç–æ–≤', index=False)
            
            # 5. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ —Ç–∏–ø–∞–º —Ü–µ—Ö–æ–≤
            print("  –°–æ–∑–¥–∞–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—Ä–µ–º–µ–Ω–∏ –ø–æ —Ç–∏–ø–∞–º —Ü–µ—Ö–æ–≤...")
            time_by_workshop_type = df.groupby('workshop_type').agg({
                'manufacturing_time_hours': 'sum',
                'total_labor_hours': 'sum',
                'workshop_name': pd.Series.nunique
            }).rename(columns={'workshop_name': 'number_of_workshops'}).reset_index()
            
            time_by_workshop_type = time_by_workshop_type.sort_values('total_labor_hours', ascending=False)
            time_by_workshop_type.to_excel(writer, sheet_name='–í—Ä–µ–º—è_–ø–æ_—Ç–∏–ø–∞–º_—Ü–µ—Ö–æ–≤', index=False)
            
            # 6. –ú–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–æ–¥—É–∫—Ç-—Ü–µ—Ö
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü—ã –ø—Ä–æ–¥—É–∫—Ç-—Ü–µ—Ö...")
            time_matrix = pd.pivot_table(
                df,
                values='manufacturing_time_hours',
                index='product_name',
                columns='workshop_name',
                aggfunc='sum',
                fill_value=0
            )
            time_matrix.to_excel(writer, sheet_name='–ú–∞—Ç—Ä–∏—Ü–∞_–ø—Ä–æ–¥—É–∫—Ç_—Ü–µ—Ö')
            
            # 7. –°–≤–æ–¥–∫–∞ –ø–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞–º
            print("  –°–æ–∑–¥–∞–Ω–∏–µ —Å–≤–æ–¥–∫–∏ –ø–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞–º...")
            material_summary = df.groupby('main_material').agg({
                'product_name': pd.Series.nunique,
                'manufacturing_time_hours': 'sum',
                'total_labor_hours': 'sum',
                'raw_material_loss_percentage': 'first'
            }).rename(columns={'product_name': 'number_of_products'}).reset_index()
            
            material_summary = material_summary.sort_values('number_of_products', ascending=False)
            material_summary.to_excel(writer, sheet_name='–°–≤–æ–¥–∫–∞_–ø–æ_–º–∞—Ç–µ—Ä–∏–∞–ª–∞–º', index=False)
            
            # 8. –≠–∫—Å–ø–æ—Ä—Ç –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
            print("  –≠–∫—Å–ø–æ—Ä—Ç –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö...")
            df.to_excel(writer, sheet_name='–ò—Å—Ö–æ–¥–Ω—ã–µ_–¥–∞–Ω–Ω—ã–µ', index=False)
        
        # –ó–∞–∫—Ä—ã–≤–∞–µ–º —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ
        conn.close()
        
        print(f"\n‚úì –î–µ—Ç–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –æ—Ç—á–µ—Ç —Å–æ–∑–¥–∞–Ω: {output_path}")
        return output_path
    
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –¥–µ—Ç–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –æ—Ç—á–µ—Ç–∞: {e}")
        if 'conn' in locals():
            conn.close()
        raise


# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–π
def main():
    """
    –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–π —ç–∫—Å–ø–æ—Ä—Ç–∞
    """
    print("=" * 60)
    print("–≠–ö–°–ü–û–†–¢ –î–ê–ù–ù–´–• –ò–ó –ë–ê–ó–´ –î–ê–ù–ù–´–• SQLite –í EXCEL")
    print("=" * 60)
    
    db_file = 'furniture_production.db'
    
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
    if not os.path.exists(db_file):
        print(f"–ë–∞–∑–∞ –¥–∞–Ω–Ω—ã—Ö '{db_file}' –Ω–µ –Ω–∞–π–¥–µ–Ω–∞.")
        print("–°–Ω–∞—á–∞–ª–∞ –∑–∞–ø—É—Å—Ç–∏—Ç–µ —Å–∫—Ä–∏–ø—Ç —Å–æ–∑–¥–∞–Ω–∏—è –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö.")
        return
    
    try:
        # –í–∞—Ä–∏–∞–Ω—Ç 1: –ü—Ä–æ—Å—Ç–æ–π —ç–∫—Å–ø–æ—Ä—Ç –≤—Å–µ—Ö —Ç–∞–±–ª–∏—Ü
        print("\n1. –ü—Ä–æ—Å—Ç–æ–π —ç–∫—Å–ø–æ—Ä—Ç –≤—Å–µ—Ö —Ç–∞–±–ª–∏—Ü...")
        export_database_to_excel(db_file, 'simple_export.xlsx')
        
        # –í–∞—Ä–∏–∞–Ω—Ç 2: –≠–∫—Å–ø–æ—Ä—Ç —Å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–º–∏ –æ—Ç—á–µ—Ç–∞–º–∏
        print("\n2. –≠–∫—Å–ø–æ—Ä—Ç —Å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–º–∏ –æ—Ç—á–µ—Ç–∞–º–∏...")
        export_with_summary_reports(db_file, 'full_export_with_reports.xlsx')
        
        # –í–∞—Ä–∏–∞–Ω—Ç 3: –≠–∫—Å–ø–æ—Ä—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ –∑–∞–ø—Ä–æ—Å–∞
        print("\n3. –≠–∫—Å–ø–æ—Ä—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ –∑–∞–ø—Ä–æ—Å–∞...")
        
        custom_query = """
        SELECT 
            product_name,
            product_type,
            article,
            minimum_partner_price,
            main_material,
            COUNT(DISTINCT workshop_name) as number_of_workshops,
            SUM(manufacturing_time_hours) as total_production_time,
            SUM(total_labor_hours) as total_labor_hours
        FROM production_data
        GROUP BY product_name
        HAVING total_production_time > 5
        ORDER BY total_production_time DESC
        """
        
        export_custom_query(db_file, custom_query, 'custom_query_result.xlsx', '–ü—Ä–æ–¥—É–∫—Ç—ã_–±–æ–ª–µ–µ_5_—á–∞—Å–æ–≤')
        
        # –í–∞—Ä–∏–∞–Ω—Ç 4: –°–æ–∑–¥–∞–Ω–∏–µ –¥–µ—Ç–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –æ—Ç—á–µ—Ç–∞
        print("\n4. –°–æ–∑–¥–∞–Ω–∏–µ –¥–µ—Ç–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –æ—Ç—á–µ—Ç–∞...")
        create_detailed_report(db_file)
        
        print("\n" + "=" * 60)
        print("–í–°–ï –û–ü–ï–†–ê–¶–ò–ò –£–°–ü–ï–®–ù–û –í–´–ü–û–õ–ù–ï–ù–´!")
        print("=" * 60)
        
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ —ç–∫—Å–ø–æ—Ä—Ç–∞: {e}")


# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ —ç–∫—Å–ø–æ—Ä—Ç–∞ (—Å–∞–º—ã–π –ø—Ä–æ—Å—Ç–æ–π –≤–∞—Ä–∏–∞–Ω—Ç)
def quick_export(db_path='furniture_production.db'):
    """
    –ë—ã—Å—Ç—Ä—ã–π —ç–∫—Å–ø–æ—Ä—Ç –≤—Å–µ—Ö –¥–∞–Ω–Ω—ã—Ö –∏–∑ –±–∞–∑—ã –≤ –æ–¥–∏–Ω Excel —Ñ–∞–π–ª
    """
    try:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        output_file = f"quick_export_{timestamp}.xlsx"
        
        print(f"–ë—ã—Å—Ç—Ä—ã–π —ç–∫—Å–ø–æ—Ä—Ç –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö '{db_path}'...")
        
        export_with_summary_reports(db_path, output_file)
        
        return output_file
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –±—ã—Å—Ç—Ä–æ–≥–æ —ç–∫—Å–ø–æ—Ä—Ç–∞: {e}")
        return None


if __name__ == "__main__":
    # –ó–∞–ø—É—Å–∫–∞–µ–º –ø—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
    main()
    
    # –ò–ª–∏ –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ —ç–∫—Å–ø–æ—Ä—Ç–∞:
    # quick_export('furniture_production.db')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
import numpy as np
from datetime import datetime
import os
import warnings
warnings.filterwarnings('ignore')

# –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Å—Ç–∏–ª—å –¥–ª—è –∫—Ä–∞—Å–∏–≤—ã—Ö –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.style.use('seaborn-v0_8-darkgrid')
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.titlesize'] = 12
matplotlib.rcParams['axes.labelsize'] = 11

def create_dashboard_pdf(csv_file_path, output_pdf_path=None):
    """
    –°–æ–∑–¥–∞–µ—Ç PDF —Ñ–∞–π–ª —Å –¥–∏–∞–≥—Ä–∞–º–º–∞–º–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –¥–∞–Ω–Ω—ã—Ö –∏–∑ CSV —Ñ–∞–π–ª–∞
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    ----------
    csv_file_path : str
        –ü—É—Ç—å –∫ CSV —Ñ–∞–π–ª—É —Å –¥–∞–Ω–Ω—ã–º–∏
    output_pdf_path : str, optional
        –ü—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è PDF —Ñ–∞–π–ª–∞. –ï—Å–ª–∏ –Ω–µ —É–∫–∞–∑–∞–Ω, –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç—Å—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏.
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    -----------
    str
        –ê–±—Å–æ–ª—é—Ç–Ω—ã–π –ø—É—Ç—å –∫ —Å–æ–∑–¥–∞–Ω–Ω–æ–º—É PDF —Ñ–∞–π–ª—É
    """
    
    try:
        # –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
        print(f"–ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Ñ–∞–π–ª–∞: {csv_file_path}")
        df = pd.read_csv(csv_file_path, encoding='utf-8-sig')
        
        print(f"–î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã: {df.shape[0]} —Å—Ç—Ä–æ–∫, {df.shape[1]} —Å—Ç–æ–ª–±—Ü–æ–≤")
        
        # –ï—Å–ª–∏ –ø—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –Ω–µ —É–∫–∞–∑–∞–Ω, –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏
        if output_pdf_path is None:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            # –ò—Å–ø–æ–ª—å–∑—É–µ–º –±–µ–∑–æ–ø–∞—Å–Ω–æ–µ –∏–º—è —Ñ–∞–π–ª–∞ (—Ç–æ–ª—å–∫–æ –ª–∞—Ç–∏–Ω–∏—Ü–∞ –∏ —Ü–∏—Ñ—Ä—ã)
            output_pdf_path = f"furniture_dashboard_{timestamp}.pdf"
        
        # –£–±–µ–¥–∏–º—Å—è, —á—Ç–æ –ø—É—Ç—å –∞–±—Å–æ–ª—é—Ç–Ω—ã–π –∏ –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–π
        output_pdf_path = os.path.abspath(output_pdf_path)
        
        # –°–æ–∑–¥–∞–µ–º –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—é, –µ—Å–ª–∏ –æ–Ω–∞ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
        output_dir = os.path.dirname(output_pdf_path)
        if output_dir and not os.path.exists(output_dir):
            os.makedirs(output_dir)
        
        print(f"PDF —Ñ–∞–π–ª –±—É–¥–µ—Ç —Å–æ—Ö—Ä–∞–Ω–µ–Ω –∫–∞–∫: {output_pdf_path}")
        
        # –°–æ–∑–¥–∞–µ–º PDF —Ñ–∞–π–ª
        with PdfPages(output_pdf_path) as pdf:
            print(f"–°–æ–∑–¥–∞–Ω–∏–µ PDF —Ñ–∞–π–ª–∞...")
            
            # –¢–∏—Ç—É–ª—å–Ω–∞—è —Å—Ç—Ä–∞–Ω–∏—Ü–∞
            fig, ax = plt.subplots(figsize=(11, 8.5))
            ax.axis('off')
            
            # –ó–∞–≥–æ–ª–æ–≤–æ–∫
            title_text = "–ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏–π –æ—Ç—á–µ—Ç –ø–æ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤—É –º–µ–±–µ–ª–∏"
            ax.text(0.5, 0.7, title_text, 
                   fontsize=24, fontweight='bold', 
                   ha='center', va='center', transform=ax.transAxes)
            
            # –ü–æ–¥–∑–∞–≥–æ–ª–æ–≤–æ–∫
            csv_filename = os.path.basename(csv_file_path)
            subtitle_text = f"–ù–∞ –æ—Å–Ω–æ–≤–µ –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Ñ–∞–π–ª–∞: {csv_filename}"
            ax.text(0.5, 0.6, subtitle_text, 
                   fontsize=14, 
                   ha='center', va='center', transform=ax.transAxes)
            
            # –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö
            info_text = f"–î–∞—Ç–∞ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏: {datetime.now().strftime('%d.%m.%Y %H:%M')}\n"
            info_text += f"–í—Å–µ–≥–æ –∑–∞–ø–∏—Å–µ–π: {df.shape[0]}\n"
            info_text += f"–í—Å–µ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤: {df['product_name'].nunique()}\n"
            info_text += f"–í—Å–µ–≥–æ —Ü–µ—Ö–æ–≤: {df['workshop_name'].nunique()}\n"
            info_text += f"–û–±—â–µ–µ –≤—Ä–µ–º—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞: {df['manufacturing_time_hours'].sum():.1f} —á–∞—Å–æ–≤\n"
            info_text += f"–û–±—â–∏–µ —Ç—Ä—É–¥–æ–∑–∞—Ç—Ä–∞—Ç—ã: {df['total_labor_hours'].sum():.1f} —á–µ–ª–æ–≤–µ–∫–æ-—á–∞—Å–æ–≤"
            
            ax.text(0.5, 0.4, info_text, 
                   fontsize=12, 
                   ha='center', va='center', transform=ax.transAxes,
                   bbox=dict(boxstyle="round,pad=0.5", facecolor="lightblue", alpha=0.5))
            
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)
            print("  ‚úì –¢–∏—Ç—É–ª—å–Ω–∞—è —Å—Ç—Ä–∞–Ω–∏—Ü–∞ —Å–æ–∑–¥–∞–Ω–∞")
            
            # 1. –ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ç–æ–≤
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã 1: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ç–æ–≤...")
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
            
            # –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –ø—Ä–æ–¥—É–∫—Ç—ã –ø–æ —Ç–∏–ø–∞–º
            product_types = df[['product_name', 'product_type']].drop_duplicates()
            type_counts = product_types['product_type'].value_counts()
            
            # –¶–≤–µ—Ç–æ–≤–∞—è –ø–∞–ª–∏—Ç—Ä–∞
            colors = plt.cm.Set3(np.linspace(0, 1, len(type_counts)))
            
            # –ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞
            wedges, texts, autotexts = ax1.pie(type_counts.values, 
                                               labels=type_counts.index,
                                               autopct='%1.1f%%',
                                               colors=colors,
                                               startangle=90,
                                               textprops={'fontsize': 10})
            
            ax1.set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ —Ç–∏–ø–∞–º', fontsize=14, fontweight='bold')
            
            # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ —Ç–∏–ø–∞–º
            bars = ax2.bar(range(len(type_counts)), type_counts.values, color=colors)
            ax2.set_xticks(range(len(type_counts)))
            ax2.set_xticklabels(type_counts.index, rotation=45, ha='right')
            ax2.set_title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ —Ç–∏–ø–∞–º', fontsize=14, fontweight='bold')
            ax2.set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤')
            
            # –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
            for bar in bars:
                height = bar.get_height()
                ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                        f'{int(height)}', ha='center', va='bottom', fontsize=10)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)
            print("  ‚úì –î–∏–∞–≥—Ä–∞–º–º–∞ 1 —Å–æ–∑–¥–∞–Ω–∞")
            
            # 2. –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞: –¢–æ–ø-10 –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã 2: –¢–æ–ø-10 –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞...")
            fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
            
            # –°—É–º–º–∞—Ä–Ω–æ–µ –≤—Ä–µ–º—è –ø–æ –ø—Ä–æ–¥—É–∫—Ç–∞–º
            product_time = df.groupby('product_name')['manufacturing_time_hours'].sum().sort_values(ascending=False)
            top_10_products = product_time.head(10)
            
            # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ç–æ–ø-10
            bars = ax1.barh(range(len(top_10_products)), top_10_products.values)
            ax1.set_yticks(range(len(top_10_products)))
            
            # –û–±—Ä–µ–∑–∞–µ–º –¥–ª–∏–Ω–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è
            labels = [name[:40] + '...' if len(name) > 40 else name for name in top_10_products.index]
            ax1.set_yticklabels(labels, fontsize=9)
            ax1.invert_yaxis()  # –°–∞–º—ã–π –±–æ–ª—å—à–æ–π –≤–≤–µ—Ä—Ö—É
            ax1.set_xlabel('–û–±—â–µ–µ –≤—Ä–µ–º—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ (—á–∞—Å—ã)')
            ax1.set_title('–¢–æ–ø-10 –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞', fontsize=14, fontweight='bold')
            
            # –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
            for i, (bar, value) in enumerate(zip(bars, top_10_products.values)):
                ax1.text(value + 0.1, bar.get_y() + bar.get_height()/2,
                        f'{value:.1f} —á', ha='left', va='center', fontsize=9)
            
            # Box plot –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ç–æ–≤
            time_by_type = df.groupby(['product_name', 'product_type'])['manufacturing_time_hours'].sum().reset_index()
            product_types_list = time_by_type['product_type'].unique()
            
            data_to_plot = [time_by_type[time_by_type['product_type'] == pt]['manufacturing_time_hours'].values 
                           for pt in product_types_list]
            
            bp = ax2.boxplot(data_to_plot, patch_artist=True)
            
            # –¶–≤–µ—Ç–∞ –¥–ª—è box plot
            for patch, color in zip(bp['boxes'], plt.cm.Pastel1(np.linspace(0, 1, len(product_types_list)))):
                patch.set_facecolor(color)
            
            ax2.set_xticklabels(product_types_list, rotation=45, ha='right')
            ax2.set_ylabel('–í—Ä–µ–º—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ (—á–∞—Å—ã)')
            ax2.set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ç–æ–≤', fontsize=14, fontweight='bold')
            ax2.grid(True, alpha=0.3)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)
            print("  ‚úì –î–∏–∞–≥—Ä–∞–º–º–∞ 2 —Å–æ–∑–¥–∞–Ω–∞")
            
            # 3. –î–∏–∞–≥—Ä–∞–º–º–∞: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Ü–µ—Ö–∞–º
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã 3: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Ü–µ—Ö–∞–º...")
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
            
            # –í—Ä–µ–º—è –ø–æ —Ü–µ—Ö–∞–º
            workshop_time = df.groupby('workshop_name')['manufacturing_time_hours'].sum().sort_values(ascending=False)
            
            # –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ —Ü–µ—Ö–∞–º
            bars = ax1.bar(range(len(workshop_time)), workshop_time.values, 
                          color=plt.cm.viridis(np.linspace(0, 1, len(workshop_time))))
            ax1.set_xticks(range(len(workshop_time)))
            ax1.set_xticklabels(workshop_time.index, rotation=90, ha='center', fontsize=8)
            ax1.set_ylabel('–û–±—â–µ–µ –≤—Ä–µ–º—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ (—á–∞—Å—ã)')
            ax1.set_title('–í—Ä–µ–º—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ –ø–æ —Ü–µ—Ö–∞–º', fontsize=14, fontweight='bold')
            
            # –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç–æ–ª–±—Ü—ã
            for bar, value in zip(bars, workshop_time.values):
                height = bar.get_height()
                ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                        f'{value:.1f}', ha='center', va='bottom', fontsize=8, rotation=0)
            
            # –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Ç–∏–ø–∞–º —Ü–µ—Ö–æ–≤
            workshop_type_time = df.groupby('workshop_type')['manufacturing_time_hours'].sum().sort_values(ascending=False)
            
            # –ö—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ –ø–æ —Ç–∏–ø–∞–º —Ü–µ—Ö–æ–≤
            wedges, texts, autotexts = ax2.pie(workshop_type_time.values, 
                                               labels=workshop_type_time.index,
                                               autopct='%1.1f%%',
                                               colors=plt.cm.Set2(np.linspace(0, 1, len(workshop_type_time))),
                                               startangle=90)
            
            ax2.set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ —Ç–∏–ø–∞–º —Ü–µ—Ö–æ–≤', fontsize=14, fontweight='bold')
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)
            print("  ‚úì –î–∏–∞–≥—Ä–∞–º–º–∞ 3 —Å–æ–∑–¥–∞–Ω–∞")
            
            # 4. –î–∏–∞–≥—Ä–∞–º–º–∞: –°—Ç–æ–∏–º–æ—Å—Ç—å vs –í—Ä–µ–º—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã 4: –°—Ç–æ–∏–º–æ—Å—Ç—å vs –í—Ä–µ–º—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞...")
            fig, ax = plt.subplots(figsize=(12, 8))
            
            # –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –ø—Ä–æ–¥—É–∫—Ç—ã —Å —Ü–µ–Ω–∞–º–∏ –∏ –≤—Ä–µ–º–µ–Ω–µ–º
            product_stats = df.groupby('product_name').agg({
                'minimum_partner_price': 'first',
                'manufacturing_time_hours': 'sum',
                'product_type': 'first'
            }).reset_index()
            
            # –¶–≤–µ—Ç–∞ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ç–æ–≤
            unique_types = product_stats['product_type'].unique()
            color_map = {pt: plt.cm.tab20(i/len(unique_types)) for i, pt in enumerate(unique_types)}
            
            # –¢–æ—á–µ—á–Ω–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞
            for pt in unique_types:
                subset = product_stats[product_stats['product_type'] == pt]
                ax.scatter(subset['manufacturing_time_hours'], 
                          subset['minimum_partner_price'],
                          s=100, 
                          alpha=0.7,
                          label=pt,
                          color=color_map[pt],
                          edgecolors='black',
                          linewidth=0.5)
            
            ax.set_xlabel('–û–±—â–µ–µ –≤—Ä–µ–º—è –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ (—á–∞—Å—ã)')
            ax.set_ylabel('–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Ü–µ–Ω–∞ –ø–∞—Ä—Ç–Ω–µ—Ä–∞ (—Ä—É–±)')
            ax.set_title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ü–µ–Ω—ã –æ—Ç –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞', fontsize=14, fontweight='bold')
            ax.legend(title='–¢–∏–ø –ø—Ä–æ–¥—É–∫—Ç–∞', fontsize=9, title_fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
            ax.grid(True, alpha=0.3)
            
            # –î–æ–±–∞–≤–ª—è–µ–º –∏–º–µ–Ω–∞ –ø—Ä–æ–¥—É–∫—Ç–æ–≤
            for idx, row in product_stats.iterrows():
                if row['manufacturing_time_hours'] > 10 or row['minimum_partner_price'] > 100000:
                    ax.annotate(row['product_name'][:20] + '...', 
                               (row['manufacturing_time_hours'], row['minimum_partner_price']),
                               fontsize=7, alpha=0.7)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)
            print("  ‚úì –î–∏–∞–≥—Ä–∞–º–º–∞ 4 —Å–æ–∑–¥–∞–Ω–∞")
            
            # 5. Heatmap: –ú–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–æ–¥—É–∫—Ç-—Ü–µ—Ö (—É–ø—Ä–æ—â–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è)
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã 5: –ú–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–æ–¥—É–∫—Ç-—Ü–µ—Ö...")
            fig, ax = plt.subplots(figsize=(14, 10))
            
            # –°–æ–∑–¥–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É
            pivot_table = pd.pivot_table(df, 
                                        values='manufacturing_time_hours',
                                        index='product_name',
                                        columns='workshop_name',
                                        aggfunc='sum',
                                        fill_value=0)
            
            # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –¥–ª—è —á–∏—Ç–∞–µ–º–æ—Å—Ç–∏
            if len(pivot_table) > 15:
                # –ë–µ—Ä–µ–º —Ç–æ–ø-15 –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ –æ–±—â–µ–º—É –≤—Ä–µ–º–µ–Ω–∏
                product_totals = pivot_table.sum(axis=1).sort_values(ascending=False)
                top_products = product_totals.head(15).index
                pivot_table = pivot_table.loc[top_products]
            
            # Heatmap
            im = ax.imshow(pivot_table.values, cmap='YlOrRd', aspect='auto')
            
            # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –æ—Å–µ–π
            ax.set_xticks(np.arange(len(pivot_table.columns)))
            ax.set_yticks(np.arange(len(pivot_table.index)))
            
            # –£–∫–æ—Ä–∞—á–∏–≤–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –¥–ª—è –æ—Å–∏ X
            x_labels = [label[:20] + '...' if len(label) > 20 else label for label in pivot_table.columns]
            ax.set_xticklabels(x_labels, rotation=90, fontsize=8)
            
            # –£–∫–æ—Ä–∞—á–∏–≤–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –¥–ª—è –æ—Å–∏ Y
            y_labels = [label[:30] + '...' if len(label) > 30 else label for label in pivot_table.index]
            ax.set_yticklabels(y_labels, fontsize=8)
            
            # –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ —è—á–µ–π–∫–∏
            for i in range(len(pivot_table.index)):
                for j in range(len(pivot_table.columns)):
                    value = pivot_table.iloc[i, j]
                    if value > 0:
                        text = ax.text(j, i, f'{value:.1f}',
                                     ha="center", va="center",
                                     color="black" if value < pivot_table.values.max()/2 else "white",
                                     fontsize=7)
            
            ax.set_title('–ú–∞—Ç—Ä–∏—Ü–∞ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞: –ü—Ä–æ–¥—É–∫—Ç √ó –¶–µ—Ö (—á–∞—Å—ã)', 
                        fontsize=14, fontweight='bold', pad=20)
            
            # –¶–≤–µ—Ç–æ–≤–∞—è —à–∫–∞–ª–∞
            cbar = plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
            cbar.set_label('–ß–∞—Å—ã –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞', fontsize=10)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)
            print("  ‚úì –î–∏–∞–≥—Ä–∞–º–º–∞ 5 —Å–æ–∑–¥–∞–Ω–∞")
            
            # 6. –î–∏–∞–≥—Ä–∞–º–º–∞: –ê–Ω–∞–ª–∏–∑ –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤
            print("  –°–æ–∑–¥–∞–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã 6: –ê–Ω–∞–ª–∏–∑ –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤...")
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
            
            # –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞–º
            material_stats = df.groupby('main_material').agg({
                'product_name': pd.Series.nunique,
                'manufacturing_time_hours': 'sum',
                'raw_material_loss_percentage': 'first'
            }).reset_index()
            
            material_stats = material_stats.sort_values('product_name', ascending=False)
            
            # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞–º
            bars1 = ax1.bar(range(len(material_stats)), material_stats['product_name'].values,
                           color=plt.cm.Paired(np.linspace(0, 1, len(material_stats))))
            ax1.set_xticks(range(len(material_stats)))
            ax1.set_xticklabels(material_stats['main_material'], rotation=45, ha='right')
            ax1.set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤')
            ax1.set_title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞–º', fontsize=14, fontweight='bold')
            
            for bar, value in zip(bars1, material_stats['product_name'].values):
                height = bar.get_height()
                ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                        f'{int(value)}', ha='center', va='bottom', fontsize=9)
            
            # –ü–æ—Ç–µ—Ä–∏ –º–∞—Ç–µ—Ä–∏–∞–ª–∞
            bars2 = ax2.bar(range(len(material_stats)), material_stats['raw_material_loss_percentage'].values * 100,
                           color=plt.cm.Set2(np.linspace(0, 1, len(material_stats))))
            ax2.set_xticks(range(len(material_stats)))
            ax2.set_xticklabels(material_stats['main_material'], rotation=45, ha='right')
            ax2.set_ylabel('–ü—Ä–æ—Ü–µ–Ω—Ç –ø–æ—Ç–µ—Ä—å –º–∞—Ç–µ—Ä–∏–∞–ª–∞ (%)')
            ax2.set_title('–ü—Ä–æ—Ü–µ–Ω—Ç –ø–æ—Ç–µ—Ä—å –ø–æ –º–∞—Ç–µ—Ä–∏–∞–ª–∞–º', fontsize=14, fontweight='bold')
            
            for bar, value in zip(bars2, material_stats['raw_material_loss_percentage'].values * 100):
                height = bar.get_height()
                ax2.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                        f'{value:.2f}%', ha='center', va='bottom', fontsize=9)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)
            print("  ‚úì –î–∏–∞–≥—Ä–∞–º–º–∞ 6 —Å–æ–∑–¥–∞–Ω–∞")
            
            # 7. –°–≤–æ–¥–Ω–∞—è —Å—Ç—Ä–∞–Ω–∏—Ü–∞ —Å —Ç–∞–±–ª–∏—Ü–∞–º–∏
            print("  –°–æ–∑–¥–∞–Ω–∏–µ —Å–≤–æ–¥–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã —Å —Ç–∞–±–ª–∏—Ü–∞–º–∏...")
            fig, axes = plt.subplots(2, 2, figsize=(14, 10))
            axes = axes.flatten()
            
            # –¢–∞–±–ª–∏—Ü–∞ 1: –¢–æ–ø-5 –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ –≤—Ä–µ–º–µ–Ω–∏
            ax = axes[0]
            ax.axis('tight')
            ax.axis('off')
            top_5_table = product_time.head().reset_index()
            top_5_table.columns = ['–ü—Ä–æ–¥—É–∫—Ç', '–í—Ä–µ–º—è (—á–∞—Å—ã)']
            top_5_table['–í—Ä–µ–º—è (—á–∞—Å—ã)'] = top_5_table['–í—Ä–µ–º—è (—á–∞—Å—ã)'].apply(lambda x: f'{x:.1f}')
            # –£–∫–æ—Ä–∞—á–∏–≤–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –ø—Ä–æ–¥—É–∫—Ç–æ–≤
            top_5_table['–ü—Ä–æ–¥—É–∫—Ç'] = top_5_table['–ü—Ä–æ–¥—É–∫—Ç'].apply(lambda x: x[:30] + '...' if len(x) > 30 else x)
            table1 = ax.table(cellText=top_5_table.values,
                             colLabels=top_5_table.columns,
                             cellLoc='center',
                             loc='center',
                             colWidths=[0.7, 0.3])
            table1.auto_set_font_size(False)
            table1.set_fontsize(9)
            table1.scale(1, 1.5)
            ax.set_title('–¢–æ–ø-5 –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞', fontsize=12, fontweight='bold', y=0.95)
            
            # –¢–∞–±–ª–∏—Ü–∞ 2: –¢–æ–ø-5 —Ü–µ—Ö–æ–≤ –ø–æ –≤—Ä–µ–º–µ–Ω–∏
            ax = axes[1]
            ax.axis('tight')
            ax.axis('off')
            top_5_workshops = workshop_time.head().reset_index()
            top_5_workshops.columns = ['–¶–µ—Ö', '–í—Ä–µ–º—è (—á–∞—Å—ã)']
            top_5_workshops['–í—Ä–µ–º—è (—á–∞—Å—ã)'] = top_5_workshops['–í—Ä–µ–º—è (—á–∞—Å—ã)'].apply(lambda x: f'{x:.1f}')
            table2 = ax.table(cellText=top_5_workshops.values,
                             colLabels=top_5_workshops.columns,
                             cellLoc='center',
                             loc='center',
                             colWidths=[0.7, 0.3])
            table2.auto_set_font_size(False)
            table2.set_fontsize(9)
            table2.scale(1, 1.5)
            ax.set_title('–¢–æ–ø-5 —Ü–µ—Ö–æ–≤ –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞', fontsize=12, fontweight='bold', y=0.95)
            
            # –¢–∞–±–ª–∏—Ü–∞ 3: –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ç–æ–≤
            ax = axes[2]
            ax.axis('tight')
            ax.axis('off')
            type_stats = df.groupby('product_type').agg({
                'product_name': pd.Series.nunique,
                'manufacturing_time_hours': 'sum',
                'minimum_partner_price': 'mean'
            }).reset_index()
            type_stats.columns = ['–¢–∏–ø', '–ö–æ–ª-–≤–æ', '–í—Ä–µ–º—è (—á)', '–°—Ä. —Ü–µ–Ω–∞']
            type_stats['–í—Ä–µ–º—è (—á)'] = type_stats['–í—Ä–µ–º—è (—á)'].apply(lambda x: f'{x:.1f}')
            type_stats['–°—Ä. —Ü–µ–Ω–∞'] = type_stats['–°—Ä. —Ü–µ–Ω–∞'].apply(lambda x: f'{x:,.0f}')
            table3 = ax.table(cellText=type_stats.values,
                             colLabels=type_stats.columns,
                             cellLoc='center',
                             loc='center')
            table3.auto_set_font_size(False)
            table3.set_fontsize(9)
            table3.scale(1, 1.5)
            ax.set_title('–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Ç–∏–ø–∞–º –ø—Ä–æ–¥—É–∫—Ç–æ–≤', fontsize=12, fontweight='bold', y=0.95)
            
            # –¢–∞–±–ª–∏—Ü–∞ 4: –û–±—â–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
            ax = axes[3]
            ax.axis('tight')
            ax.axis('off')
            
            total_stats_data = [
                ['–í—Å–µ–≥–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤', f"{df['product_name'].nunique()}"],
                ['–í—Å–µ–≥–æ —Ü–µ—Ö–æ–≤', f"{df['workshop_name'].nunique()}"],
                ['–û–±—â–µ–µ –≤—Ä–µ–º—è', f"{df['manufacturing_time_hours'].sum():.1f} —á"],
                ['–¢—Ä—É–¥–æ–∑–∞—Ç—Ä–∞—Ç—ã', f"{df['total_labor_hours'].sum():.1f} —á-—á"],
                ['–°—Ä. –≤—Ä–µ–º—è –Ω–∞ –ø—Ä–æ–¥—É–∫—Ç', f"{df.groupby('product_name')['manufacturing_time_hours'].sum().mean():.1f} —á"],
                ['–°—Ä. —Ü–µ–Ω–∞', f"{df['minimum_partner_price'].mean():,.0f} —Ä—É–±"],
                ['–ú–∏–Ω. —Ü–µ–Ω–∞', f"{df['minimum_partner_price'].min():,.0f} —Ä—É–±"],
                ['–ú–∞–∫—Å. —Ü–µ–Ω–∞', f"{df['minimum_partner_price'].max():,.0f} —Ä—É–±"]
            ]
            
            table4 = ax.table(cellText=total_stats_data,
                             cellLoc='left',
                             loc='center',
                             colWidths=[0.6, 0.4])
            table4.auto_set_font_size(False)
            table4.set_fontsize(9)
            table4.scale(1, 1.5)
            ax.set_title('–û–±—â–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞', fontsize=12, fontweight='bold', y=0.95)
            
            plt.suptitle('–°–≤–æ–¥–Ω—ã–µ —Ç–∞–±–ª–∏—Ü—ã', fontsize=16, fontweight='bold', y=0.98)
            plt.tight_layout(rect=[0, 0, 1, 0.96])
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)
            print("  ‚úì –°–≤–æ–¥–Ω–∞—è —Å—Ç—Ä–∞–Ω–∏—Ü–∞ —Å–æ–∑–¥–∞–Ω–∞")
            
            # –°—Ç—Ä–∞–Ω–∏—Ü–∞ —Å –º–µ—Ç–∞–¥–∞–Ω–Ω—ã–º–∏
            fig, ax = plt.subplots(figsize=(11, 8.5))
            ax.axis('off')
            
            metadata_text = "–ú–µ—Ç–∞–¥–∞–Ω–Ω—ã–µ –æ—Ç—á–µ—Ç–∞\n\n"
            metadata_text += f"–ò—Å—Ö–æ–¥–Ω—ã–π —Ñ–∞–π–ª: {os.path.basename(csv_file_path)}\n"
            metadata_text += f"–î–∞—Ç–∞ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏: {datetime.now().strftime('%d.%m.%Y %H:%M:%S')}\n"
            metadata_text += f"–í—Å–µ–≥–æ —Å—Ç—Ä–∞–Ω–∏—Ü –≤ –æ—Ç—á–µ—Ç–µ: {pdf.get_pagecount()}\n\n"
            metadata_text += f"–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–∞–Ω–Ω—ã—Ö:\n"
            metadata_text += f"‚Ä¢ –í—Å–µ–≥–æ –∑–∞–ø–∏—Å–µ–π: {df.shape[0]}\n"
            metadata_text += f"‚Ä¢ –°—Ç–æ–ª–±—Ü–æ–≤: {df.shape[1]}\n"
            metadata_text += f"‚Ä¢ –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤: {df['product_name'].nunique()}\n"
            metadata_text += f"‚Ä¢ –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ü–µ—Ö–æ–≤: {df['workshop_name'].nunique()}\n"
            metadata_text += f"‚Ä¢ –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ç–∏–ø–æ–≤ –ø—Ä–æ–¥—É–∫—Ç–æ–≤: {df['product_type'].nunique()}\n"
            metadata_text += f"‚Ä¢ –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤: {df['main_material'].nunique()}\n\n"
            metadata_text += "–°–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –æ—Ç—á–µ—Ç–∞:\n"
            metadata_text += "1. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ —Ç–∏–ø–∞–º\n"
            metadata_text += "2. –¢–æ–ø-10 –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞\n"
            metadata_text += "3. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Ü–µ—Ö–∞–º\n"
            metadata_text += "4. –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Ü–µ–Ω—ã –æ—Ç –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞\n"
            metadata_text += "5. –ú–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–æ–¥—É–∫—Ç-—Ü–µ—Ö\n"
            metadata_text += "6. –ê–Ω–∞–ª–∏–∑ –º–∞—Ç–µ—Ä–∏–∞–ª–æ–≤\n"
            metadata_text += "7. –°–≤–æ–¥–Ω—ã–µ —Ç–∞–±–ª–∏—Ü—ã\n\n"
            metadata_text += "–û—Ç—á–µ—Ç —Å–æ–∑–¥–∞–Ω –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º:\n"
            metadata_text += "‚Ä¢ Python pandas –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö\n"
            metadata_text += "‚Ä¢ Matplotlib –∏ Seaborn –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏\n"
            
            ax.text(0.1, 0.9, metadata_text, 
                   fontsize=12, 
                   va='top', transform=ax.transAxes,
                   bbox=dict(boxstyle="round,pad=1", facecolor="lightgray", alpha=0.3))
            
            pdf.savefig(fig, bbox_inches='tight')
            plt.close(fig)
            print("  ‚úì –°—Ç—Ä–∞–Ω–∏—Ü–∞ —Å –º–µ—Ç–∞–¥–∞–Ω–Ω—ã–º–∏ —Å–æ–∑–¥–∞–Ω–∞")
        
        print(f"\n{'='*60}")
        print(f"‚úì –û–¢–ß–ï–¢ –£–°–ü–ï–®–ù–û –°–û–ó–î–ê–ù!")
        print(f"{'='*60}")
        print(f"–§–∞–π–ª: {output_pdf_path}")
        print(f"–í—Å–µ–≥–æ —Å—Ç—Ä–∞–Ω–∏—Ü: 8")
        print(f"–î–∞—Ç–∞ —Å–æ–∑–¥–∞–Ω–∏—è: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        
        # –û—Ç–∫—Ä—ã–≤–∞–µ–º —Ñ–∞–π–ª (—Ä–∞–±–æ—Ç–∞–µ—Ç –Ω–∞ Windows, Mac, Linux)
        try:
            import platform
            import subprocess
            
            system = platform.system()
            if system == 'Windows':
                os.startfile(output_pdf_path)
            elif system == 'Darwin':  # macOS
                subprocess.run(['open', output_pdf_path])
            else:  # Linux –∏ –¥—Ä—É–≥–∏–µ
                subprocess.run(['xdg-open', output_pdf_path])
            print(f"‚úì PDF —Ñ–∞–π–ª –æ—Ç–∫—Ä—ã—Ç –≤ –ø—Ä–æ–≥—Ä–∞–º–º–µ –¥–ª—è –ø—Ä–æ—Å–º–æ—Ç—Ä–∞")
        except:
            print(f"‚ö† –ù–µ —É–¥–∞–ª–æ—Å—å –æ—Ç–∫—Ä—ã—Ç—å —Ñ–∞–π–ª –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏. –û—Ç–∫—Ä–æ–π—Ç–µ –≤—Ä—É—á–Ω—É—é: {output_pdf_path}")
        
        return output_pdf_path
    
    except FileNotFoundError:
        print(f"–û—à–∏–±–∫–∞: –§–∞–π–ª {csv_file_path} –Ω–µ –Ω–∞–π–¥–µ–Ω")
        raise
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –æ—Ç—á–µ—Ç–∞: {str(e)}")
        raise


def create_simple_pdf_report(csv_file_path, output_dir=None):
    """
    –°–æ–∑–¥–∞–µ—Ç —É–ø—Ä–æ—â–µ–Ω–Ω—ã–π PDF –æ—Ç—á–µ—Ç –∏ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç –µ–≥–æ –≤ —Ç–µ–∫—É—â–µ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    ----------
    csv_file_path : str
        –ü—É—Ç—å –∫ CSV —Ñ–∞–π–ª—É —Å –¥–∞–Ω–Ω—ã–º–∏
    output_dir : str, optional
        –î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è PDF —Ñ–∞–π–ª–∞
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    -----------
    str
        –ü—É—Ç—å –∫ —Å–æ–∑–¥–∞–Ω–Ω–æ–º—É PDF —Ñ–∞–π–ª—É
    """
    
    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—é –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è
    if output_dir is None:
        output_dir = os.getcwd()  # –¢–µ–∫—É—â–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è
    elif not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_pdf_path = os.path.join(output_dir, f"report_{timestamp}.pdf")
    
    print(f"–°–æ–∑–¥–∞–Ω–∏–µ —É–ø—Ä–æ—â–µ–Ω–Ω–æ–≥–æ –æ—Ç—á–µ—Ç–∞...")
    print(f"–ò—Å—Ö–æ–¥–Ω—ã–π —Ñ–∞–π–ª: {csv_file_path}")
    print(f"–í—ã—Ö–æ–¥–Ω–æ–π —Ñ–∞–π–ª: {output_pdf_path}")
    
    return create_dashboard_pdf(csv_file_path, output_pdf_path)


def fix_and_create_report():
    """
    –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–∏—è –æ—à–∏–±–æ–∫ –∏ —Å–æ–∑–¥–∞–Ω–∏—è –æ—Ç—á–µ—Ç–∞
    """
    
    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –ø—É—Ç—å –∫ CSV —Ñ–∞–π–ª—É
    csv_file = 'combined_data.csv'
    
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞
    if not os.path.exists(csv_file):
        print(f"–û—à–∏–±–∫–∞: –§–∞–π–ª {csv_file} –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ —Ç–µ–∫—É—â–µ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏")
        print(f"–¢–µ–∫—É—â–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è: {os.getcwd()}")
        print(f"–°–æ–¥–µ—Ä–∂–∏–º–æ–µ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏:")
        for file in os.listdir('.'):
            print(f"  - {file}")
        return None
    
    try:
        # –°–æ–∑–¥–∞–µ–º –æ—Ç—á–µ—Ç –≤ —Ç–µ–∫—É—â–µ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏ —Å –±–µ–∑–æ–ø–∞—Å–Ω—ã–º –∏–º–µ–Ω–µ–º
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        output_file = f"furniture_report_{timestamp}.pdf"
        
        print(f"–°–æ–∑–¥–∞–Ω–∏–µ –æ—Ç—á–µ—Ç–∞ –∏–∑ —Ñ–∞–π–ª–∞: {csv_file}")
        print(f"–û—Ç—á–µ—Ç –±—É–¥–µ—Ç —Å–æ—Ö—Ä–∞–Ω–µ–Ω –∫–∞–∫: {output_file}")
        
        # –ó–∞–ø—É—Å–∫–∞–µ–º —Å–æ–∑–¥–∞–Ω–∏–µ –æ—Ç—á–µ—Ç–∞
        pdf_path = create_dashboard_pdf(csv_file, output_file)
        
        print(f"\n‚úÖ –û—Ç—á–µ—Ç —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω!")
        print(f"üìÑ –§–∞–π–ª: {pdf_path}")
        print(f"üìÅ –ü–æ–ª–Ω—ã–π –ø—É—Ç—å: {os.path.abspath(pdf_path)}")
        
        return pdf_path
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –æ—Ç—á–µ—Ç–∞: {e}")
        return None


# –ï—Å–ª–∏ —Å–∫—Ä–∏–ø—Ç –∑–∞–ø—É—Å–∫–∞–µ—Ç—Å—è –Ω–∞–ø—Ä—è–º—É—é
if __name__ == "__main__":
    print("=" * 60)
    print("–ì–ï–ù–ï–†–ê–¢–û–† –û–¢–ß–ï–¢–û–í –ü–û –ü–†–û–ò–ó–í–û–î–°–¢–í–£ –ú–ï–ë–ï–õ–ò")
    print("=" * 60)
    
    # –°–æ–∑–¥–∞–µ–º –æ—Ç—á–µ—Ç
    pdf_file = fix_and_create_report()
    
    if pdf_file:
        print(f"\nüìä –û—Ç—á–µ—Ç —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω: {pdf_file}")
        print(f"üìç –§–∞–π–ª —Å–æ—Ö—Ä–∞–Ω–µ–Ω –≤: {os.path.abspath(pdf_file)}")
        
        # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –∏–Ω—Å—Ç—Ä—É–∫—Ü–∏—é –ø–æ –æ—Ç–∫—Ä—ã—Ç–∏—é —Ñ–∞–π–ª–∞
        print("\n" + "=" * 60)
        print("–ò–ù–°–¢–†–£–ö–¶–ò–Ø –ü–û –û–¢–ö–†–´–¢–ò–Æ –§–ê–ô–õ–ê:")
        print("=" * 60)
        print("1. –ù–∞–π–¥–∏—Ç–µ —Ñ–∞–π–ª –≤ –ø—Ä–æ–≤–æ–¥–Ω–∏–∫–µ:")
        print(f"   {os.path.abspath(pdf_file)}")
        print("\n2. –û—Ç–∫—Ä–æ–π—Ç–µ —Ñ–∞–π–ª –¥–≤–æ–π–Ω—ã–º –∫–ª–∏–∫–æ–º")
        print("\n3. –ò–ª–∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–º–∞–Ω–¥—É –≤ —Ç–µ—Ä–º–∏–Ω–∞–ª–µ:")
        print(f"   {pdf_file}")
    else:
        print("–ù–µ —É–¥–∞–ª–æ—Å—å —Å–æ–∑–¥–∞—Ç—å –æ—Ç—á–µ—Ç")
```
